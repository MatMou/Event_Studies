Sessions 3 through 5 answered a single question: is the average abnormal return statistically different from zero? This question is important, but it treats all firms in the sample as homogeneous. In practice, the abnormal return varies across firms, and understanding *why* it varies is often the central research question.

Cross-sectional regression analysis extends the event study by modelling the firm-level CAR as a function of observable characteristics. The seminal papers in this tradition include Asquith (1983), who regressed target CARs on deal characteristics in mergers, and Kothari and Warner (2007), who formalized the two-step procedure (estimate CARs in the first step, regress CARs on explanatory variables in the second step) that is now standard.

This session develops the cross-sectional regression framework from the ground up. We start with the econometric specification, discuss the assumptions, and derive the correct standard errors. We then apply the method to a new dataset: a sample of merger and acquisition (M&A) announcements involving publicly traded target firms. This application is canonical in corporate finance and illustrates the key issues: heteroskedasticity in the CARs, the role of deal characteristics (payment method, deal size, bidder competition), and the interpretation of coefficient estimates as marginal effects on shareholder wealth.

## 1. From Aggregate Inference to Cross-Sectional Explanation

The event study methodology developed in Sessions 1 through 5 produces, for each firm $i$, a cumulative abnormal return $CAR_i(\tau_a, \tau_b)$ over a chosen event window. The tests of Sessions 4 and 5 ask whether the cross-sectional average, the CAAR, is different from zero. But the individual CARs are typically dispersed around the CAAR: some firms show large positive abnormal returns, others show small or negative ones.

Cross-sectional regression asks what drives this dispersion. The general model is:

$$
CAR_i = \gamma_0 + \gamma_1 X_{1,i} + \gamma_2 X_{2,i} + \cdots + \gamma_K X_{K,i} + \eta_i \tag{1}
$$

where $CAR_i$ is the cumulative abnormal return for firm $i$, $X_{1,i}, \ldots, X_{K,i}$ are observable firm or event characteristics measured before or at the time of the event, and $\eta_i$ is an error term capturing unobserved heterogeneity.

The coefficients $\gamma_1, \ldots, \gamma_K$ measure the marginal effect of each characteristic on the abnormal return, holding the others constant. The intercept $\gamma_0$ captures the average CAR for a firm at the reference point of all explanatory variables.

This model is estimated by OLS. The key econometric issues are as follows.

**Heteroskedasticity.** The CAR is estimated with error from the first stage (the market model), and the precision of the estimate varies across firms. Firms with higher return volatility have noisier CARs. This means that $\text{Var}(\eta_i)$ is not constant across firms, violating the homoskedasticity assumption of OLS. The coefficients remain unbiased and consistent, but the standard OLS standard errors are incorrect. The solution is to use heteroskedasticity-consistent (HC) standard errors (White, 1980). We use the HC1 variant, which applies a degrees-of-freedom correction.

**Generated regressor problem.** The dependent variable $CAR_i$ is itself an estimate, not an observation. This means it contains estimation error from the first stage. If the first-stage estimation error is uncorrelated with the explanatory variables (which holds when the explanatory variables are predetermined), OLS remains consistent. However, the standard errors should in principle account for the first-stage estimation uncertainty. In practice, this correction is small when the estimation window is long (say, $L > 100$ trading days), because the first-stage estimation error is small relative to the cross-sectional variation in CARs. We note this caveat but follow the standard practice of treating the CARs as observed.

**Endogeneity.** If an explanatory variable is correlated with the error term $\eta_i$, the OLS estimate of its coefficient is biased. For example, if deal size and unobserved bidder quality are positively correlated, the coefficient on deal size captures both the direct effect and the confounding effect of quality. This is a standard omitted variable problem, and the solutions (instrumental variables, control variables, matching) are the same as in any cross-sectional regression. We discuss this further in Section 10.

**Multicollinearity.** When explanatory variables are highly correlated with each other, the individual coefficient estimates become imprecise (large standard errors), even though the regression as a whole may have good fit. The variance inflation factor (VIF) is the standard diagnostic: $VIF_k = 1 / (1 - R^2_k)$, where $R^2_k$ is the $R^2$ from regressing $X_k$ on all other explanatory variables. A VIF above 10 indicates severe collinearity.

## 2. The Two-Step Procedure

The standard cross-sectional event study proceeds in two steps:

**Step 1: Estimate CARs.** For each firm $i$, estimate the market model over the estimation window and compute the cumulative abnormal return $CAR_i(\tau_a, \tau_b)$ over the event window, as in Sessions 1 through 3. This step also produces the residual standard deviation $\hat{\sigma}_{\varepsilon_i}$, which will be used as a precision weight.

**Step 2: Cross-sectional regression.** Regress the vector of CARs on the matrix of explanatory variables:

$$
\mathbf{CAR} = \mathbf{X} \boldsymbol{\gamma} + \boldsymbol{\eta}
$$

where $\mathbf{CAR}$ is $N \times 1$, $\mathbf{X}$ is $N \times (K+1)$ (including the constant), $\boldsymbol{\gamma}$ is $(K+1) \times 1$, and $\boldsymbol{\eta}$ is $N \times 1$.

The OLS estimator is:

$$
\hat{\boldsymbol{\gamma}} = (\mathbf{X}'\mathbf{X})^{-1} \mathbf{X}' \mathbf{CAR}
$$

The heteroskedasticity-consistent covariance matrix (White, 1980) is:

$$
\widehat{\text{Var}}(\hat{\boldsymbol{\gamma}}) = (\mathbf{X}'\mathbf{X})^{-1} \left( \sum_{i=1}^{N} \hat{\eta}_i^2 \mathbf{x}_i \mathbf{x}_i' \right) (\mathbf{X}'\mathbf{X})^{-1}
$$

where $\hat{\eta}_i = CAR_i - \mathbf{x}_i' \hat{\boldsymbol{\gamma}}$ are the OLS residuals and $\mathbf{x}_i$ is the $i$-th row of $\mathbf{X}$ (as a column vector).

An alternative to HC standard errors is **weighted least squares (WLS)**, where each observation is weighted by the inverse of its estimated variance. If we use $w_i = 1 / \hat{\sigma}^2_{\varepsilon_i}$ as weights, firms with more precise CARs receive higher weight. This is efficient under the assumption that the variance of $\eta_i$ is proportional to $\hat{\sigma}^2_{\varepsilon_i}$, which is plausible when the cross-sectional variation in CARs is driven primarily by estimation noise. In practice, WLS and OLS with HC standard errors give similar results when the sample is reasonably large and the explanatory variables are not correlated with firm-level volatility.

## 3. Application: Merger and Acquisition Announcements

### 3.1 Why M&A?

The cross-sectional determinants of takeover announcement returns are one of the most studied topics in corporate finance. The key empirical regularities, established over decades of research, are the following (Betton, Eckbo, and Thorburn, 2008; Eckbo, 2009):

Target shareholders earn large positive abnormal returns on announcement, typically in the range of 15 to 30 percent. The magnitude depends on deal characteristics: cash offers pay higher premiums than stock offers (because cash offers are taxable and require a premium to compensate target shareholders); competitive bids (multiple bidders) produce higher target returns than uncontested bids; hostile offers pay higher premiums than friendly negotiations; and smaller targets tend to earn higher percentage returns (though lower dollar returns).

Acquirer shareholders, in contrast, earn returns that are approximately zero on average, and often slightly negative for large acquirers (the "hubris hypothesis" of Roll, 1986, and the "winner's curse" in competitive bidding).

These regularities make M&A announcements an ideal application for cross-sectional regression: the CARs are large (high signal-to-noise ratio), the cross-sectional variation is substantial, and the explanatory variables (payment method, deal size, competition) are readily observable and have clear theoretical predictions.

### 3.2 Sample Construction

We construct a sample of recent M&A announcements involving publicly traded U.S. target firms. For each deal, we record the announcement date, the target ticker, the deal value, the payment method (cash, stock, or mixed), whether the deal was contested (multiple bidders), and the target's market capitalization and sector. In a research setting, this data would come from the SDC Platinum, Refinitiv, or Bloomberg M&A databases. For this session, we hand-collect a sample of well-known deals from public sources.

The choice of event window for cross-sectional regression is important. A short window (e.g., $[-1, +1]$) captures the immediate market reaction to the announcement and minimizes contamination from confounding events. A longer window (e.g., $[-5, +5]$ or $[-10, +10]$) may capture pre-announcement run-up (leakage, rumors) and post-announcement price adjustment, but at the cost of introducing more noise. We compute CARs for multiple windows and report regression results for the $[-1, +1]$ window as the primary specification, with longer windows as robustness checks.

```python
import pandas as pd
import numpy as np
import yfinance as yf
import statsmodels.api as sm
import matplotlib.pyplot as plt
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# -- M&A Sample: publicly announced deals with US-listed targets --
# Each row: target ticker, announcement date, deal value ($B), payment method,
#           whether contested (multiple bidders), target sector.

deals = pd.DataFrame({
    'target': ['ATVI', 'VMW', 'SGEN', 'SAVE', 'IRBT', 'FORG', 'MBLY', 'NATI', 'TSEM', 'GPN'],
    'announcement': pd.to_datetime([
        '2022-01-18',  # Microsoft-Activision
        '2022-05-26',  # Broadcom-VMware
        '2023-03-13',  # Pfizer-Seagen
        '2022-02-07',  # Frontier-Spirit Airlines
        '2022-08-05',  # Amazon-iRobot
        '2022-10-11',  # Thoma Bravo-ForgeRock
        '2022-03-07',  # Intel-Mobileye (re-IPO context, announced earlier)
        '2023-04-27',  # Emerson-National Instruments
        '2022-02-15',  # Intel-Tower Semiconductor
        '2023-08-08',  # Global Payments (GTCR deal)
    ]),
    'acquirer_name': ['Microsoft', 'Broadcom', 'Pfizer', 'Frontier Airlines',
                      'Amazon', 'Thoma Bravo', 'Intel', 'Emerson Electric',
                      'Intel', 'GTCR'],
    'deal_value_bn': [68.7, 61.0, 43.0, 6.6, 1.7, 2.3, 5.4, 8.2, 5.4, 4.6],
    'payment': ['Cash', 'Cash+Stock', 'Cash', 'Stock', 'Cash', 'Cash',
                'Cash', 'Cash', 'Cash', 'Cash'],
    'contested': [False, False, False, True, False, False, False, False, False, False],
    'sector': ['Tech', 'Tech', 'Healthcare', 'Airlines', 'Consumer',
               'Tech', 'Tech', 'Industrial', 'Tech', 'Fintech'],
})

print("M&A Sample")
print("=" * 85)
print(deals[['target', 'acquirer_name', 'announcement', 'deal_value_bn', 'payment', 'contested']].to_string(index=False))
print(f"\n{len(deals)} deals, {deals['payment'].value_counts().to_dict()}")
```

## 4. Step 1: Estimating Target CARs

We estimate the market model for each target firm over a 250-day estimation window ending 20 days before the announcement. We use a longer buffer (20 days rather than the 10 used in Sessions 2-5) because M&A rumors and pre-announcement run-up are common, and we want to ensure the estimation window is free of event contamination.

For each target, we compute CARs for three event windows: $[-1, +1]$ (immediate reaction), $[-5, +5]$ (short-term, captures some run-up), and $[-10, +10]$ (wider window). The primary analysis uses $[-1, +1]$.

```python
# -- Download data and estimate market model --
market_ticker = '^GSPC'
start_date = '2020-01-01'
end_date = '2024-01-01'

data_market = yf.download(market_ticker, start=start_date, end=end_date, progress=False)
if isinstance(data_market.columns, pd.MultiIndex):
    data_market.columns = data_market.columns.get_level_values(0)
market_prices = data_market['Close']

target_tickers = deals['target'].tolist()
data_targets = yf.download(target_tickers, start=start_date, end=end_date, progress=False)
if isinstance(data_targets.columns, pd.MultiIndex):
    target_prices = data_targets['Close']
else:
    target_prices = data_targets[['Close']].rename(columns={'Close': target_tickers[0]})

market_returns = market_prices.pct_change().dropna()
target_returns = target_prices.pct_change().dropna()

common_idx = market_returns.index.intersection(target_returns.index)
market_returns = market_returns.loc[common_idx]
target_returns = target_returns.loc[common_idx]
trading_days = market_returns.index

print(f"Market data: {len(common_idx)} trading days ({common_idx[0].date()} to {common_idx[-1].date()})")
print(f"Targets downloaded: {target_returns.shape[1]}")
```

```python
# -- Estimate market model and compute CARs for each target --
est_window_length = 250
buffer = 20
event_windows = [(-1, 1), (-5, 5), (-10, 10)]

results_list = []

for _, deal in deals.iterrows():
    tick = deal['target']
    edate = deal['announcement']

    if tick not in target_returns.columns:
        print(f"  {tick}: not in downloaded data, skipping")
        continue

    event_idx = trading_days.get_indexer([edate], method='ffill')[0]

    # Estimation window
    est_end = event_idx - buffer - 1
    est_start = est_end - est_window_length + 1

    if est_start < 0:
        print(f"  {tick}: insufficient estimation data, skipping")
        continue

    y_est = target_returns.iloc[est_start:est_end + 1][tick].values
    x_est = market_returns.iloc[est_start:est_end + 1].values
    valid = ~(np.isnan(y_est) | np.isnan(x_est))
    y_clean, x_clean = y_est[valid], x_est[valid]

    if len(y_clean) < 60:
        print(f"  {tick}: too few valid obs ({len(y_clean)}), skipping")
        continue

    X_est = sm.add_constant(x_clean)
    ols = sm.OLS(y_clean, X_est).fit()
    alpha_hat, beta_hat = ols.params[0], ols.params[1]
    sigma_hat = np.sqrt(ols.mse_resid)

    row = {
        'target': tick,
        'acquirer': deal['acquirer_name'],
        'date': edate,
        'deal_value_bn': deal['deal_value_bn'],
        'payment': deal['payment'],
        'contested': deal['contested'],
        'sector': deal['sector'],
        'alpha': alpha_hat,
        'beta': beta_hat,
        'sigma': sigma_hat,
        'R2': ols.rsquared,
    }

    # Compute CARs for each event window
    for tau_a, tau_b in event_windows:
        ev_start = event_idx + tau_a
        ev_end = event_idx + tau_b
        if ev_end >= len(trading_days):
            row[f'CAR_[{tau_a:+d},{tau_b:+d}]'] = np.nan
            continue

        y_evt = target_returns.iloc[ev_start:ev_end + 1][tick].values
        x_evt = market_returns.iloc[ev_start:ev_end + 1].values
        ar = y_evt - (alpha_hat + beta_hat * x_evt)
        row[f'CAR_[{tau_a:+d},{tau_b:+d}]'] = np.nansum(ar)

    results_list.append(row)

results_df = pd.DataFrame(results_list)

# Display CARs
print("\nEstimated Target CARs (%)")
print("=" * 90)
display_cols = ['target', 'acquirer', 'deal_value_bn', 'payment']
for tau_a, tau_b in event_windows:
    col = f'CAR_[{tau_a:+d},{tau_b:+d}]'
    display_cols.append(col)

temp = results_df[display_cols].copy()
for tau_a, tau_b in event_windows:
    col = f'CAR_[{tau_a:+d},{tau_b:+d}]'
    temp[col] = temp[col] * 100
print(temp.to_string(index=False, float_format='{:.2f}'.format))

print(f"\nAverage CAR [-1,+1]: {results_df['CAR_[-1,+1]'].mean()*100:.2f}%")
print(f"Average CAR [-5,+5]: {results_df['CAR_[-5,+5]'].mean()*100:.2f}%")
```

## 6. Constructing the Regression Variables

The explanatory variables in a cross-sectional event study regression should satisfy two conditions: they must be observable at or before the event date (no look-ahead bias), and they should have theoretical motivation.

For M&A target returns, the standard variables in the literature are:

**Payment method.** Cash offers require the target shareholders to realize capital gains immediately (taxable event) and eliminate the upside if the combined firm outperforms. Stock offers allow deferral of capital gains and give target shareholders a share in the synergies. The tax and risk-sharing arguments predict that cash offers pay a higher premium. We encode this as a dummy variable $D_{cash,i}$ equal to one for all-cash deals.

**Deal size (log).** Larger deals attract more scrutiny from regulators and shareholders, and may involve more complex negotiations. The effect on target returns is ambiguous: larger deals may signal higher synergies (positive) or may reflect empire-building by the acquirer (which is an acquirer problem, not a target problem). We use the natural logarithm of deal value in billions, $\ln(\text{DealValue}_i)$, to reduce the influence of very large deals.

**Contested bid.** When multiple bidders compete for the target, the target's bargaining power increases and the premium is bid up. The prediction is unambiguous: $D_{contested,i} = 1$ should be associated with a higher CAR.

**Target beta.** The target's market model beta captures its systematic risk exposure. High-beta targets are more volatile, which may lead to noisier CARs. Including beta also controls for the possibility that the abnormal return is partly driven by omitted risk factors.

**Target residual volatility.** The estimation-window $\hat{\sigma}_{\varepsilon_i}$ measures the target's idiosyncratic risk. Firms with higher idiosyncratic volatility have noisier CARs, and the coefficient on volatility captures whether noisier firms systematically experience different announcement returns (e.g., because they are smaller or less followed by analysts).

```python
# -- Construct regression variables --

reg_df = results_df.copy()

# Dependent variable: CAR [-1,+1] in percentage points
reg_df['CAR_pct'] = reg_df['CAR_[-1,+1]'] * 100

# Explanatory variables
reg_df['cash_dummy'] = (reg_df['payment'] == 'Cash').astype(int)
reg_df['ln_deal_value'] = np.log(reg_df['deal_value_bn'])
reg_df['contested_dummy'] = reg_df['contested'].astype(int)
reg_df['target_beta'] = reg_df['beta']
reg_df['target_sigma'] = reg_df['sigma'] * 100  # in percentage points for readability

print("Regression Variables: Summary Statistics")
print("=" * 70)
stats_cols = ['CAR_pct', 'cash_dummy', 'ln_deal_value', 'contested_dummy', 'target_beta', 'target_sigma']
stats_table = reg_df[stats_cols].describe().T[['mean', 'std', 'min', '50%', 'max']]
stats_table.columns = ['Mean', 'Std', 'Min', 'Median', 'Max']
print(stats_table.to_string(float_format='{:.3f}'.format))

print(f"\nPayment breakdown: {reg_df['payment'].value_counts().to_dict()}")
print(f"Contested deals: {reg_df['contested'].sum()} out of {len(reg_df)}")
```

## 7. The Baseline OLS Regression

We begin with the univariate regression of $CAR_i$ on the cash payment dummy, which tests the most robust prediction in the M&A literature:

$$
CAR_i = \gamma_0 + \gamma_1 D_{cash,i} + \eta_i \tag{2}
$$

We then extend to the full multivariate specification:

$$
CAR_i = \gamma_0 + \gamma_1 D_{cash,i} + \gamma_2 \ln(\text{DealValue}_i) + \gamma_3 D_{contested,i} + \gamma_4 \beta_i + \gamma_5 \hat{\sigma}_{\varepsilon_i} + \eta_i \tag{3}
$$

All regressions use heteroskedasticity-consistent (HC1) standard errors. We report the OLS coefficients, robust standard errors, t-statistics, and p-values, alongside the $R^2$ and adjusted $R^2$.

```python
# -- OLS Regressions with HC1 standard errors --

# Remove rows with missing CARs
reg_clean = reg_df.dropna(subset=['CAR_pct']).copy()

# Model 1: Univariate (cash dummy only)
y = reg_clean['CAR_pct']
X1 = sm.add_constant(reg_clean[['cash_dummy']])
model1 = sm.OLS(y, X1).fit(cov_type='HC1')

# Model 2: Core deal characteristics
X2 = sm.add_constant(reg_clean[['cash_dummy', 'ln_deal_value', 'contested_dummy']])
model2 = sm.OLS(y, X2).fit(cov_type='HC1')

# Model 3: Full specification
X3 = sm.add_constant(reg_clean[['cash_dummy', 'ln_deal_value', 'contested_dummy',
                                  'target_beta', 'target_sigma']])
model3 = sm.OLS(y, X3).fit(cov_type='HC1')

# Display results
from statsmodels.iolib.summary2 import summary_col
comparison = summary_col([model1, model2, model3],
                         model_names=['(1) Univariate', '(2) Deal Chars.', '(3) Full'],
                         stars=True,
                         info_dict={'N': lambda x: f'{int(x.nobs)}',
                                    'R-squared': lambda x: f'{x.rsquared:.3f}',
                                    'Adj. R-sq': lambda x: f'{x.rsquared_adj:.3f}',
                                    'F-stat': lambda x: f'{x.fvalue:.2f}'})
print("Cross-Sectional Regression: CAR [-1,+1] (%)")
print("=" * 70)
print("Dependent variable: CAR [-1,+1] in percentage points")
print("Standard errors: heteroskedasticity-consistent (HC1)")
print()
print(comparison)
```

## 8. Weighted Least Squares

As discussed in Section 2, the CARs have heterogeneous precision: firms with lower residual standard deviation $\hat{\sigma}_{\varepsilon_i}$ have more precisely estimated CARs. Weighted least squares (WLS) exploits this information by assigning higher weight to more precisely estimated observations.

The WLS estimator solves:

$$
\hat{\boldsymbol{\gamma}}_{WLS} = (\mathbf{X}' \mathbf{W} \mathbf{X})^{-1} \mathbf{X}' \mathbf{W} \, \mathbf{CAR}
$$

where $\mathbf{W} = \text{diag}(w_1, \ldots, w_N)$ and $w_i = 1 / \hat{\sigma}^2_{\varepsilon_i}$.

Under the assumption that $\text{Var}(\eta_i) = c \cdot \hat{\sigma}^2_{\varepsilon_i}$ for some constant $c$, WLS is the efficient (minimum-variance) estimator. This assumption is reasonable when the dominant source of cross-sectional variation in CARs is estimation noise rather than genuine heterogeneity in the event effect. When both are present, WLS may be less efficient than OLS, but it remains consistent.

We re-estimate the full model (3) using WLS and compare the coefficients and standard errors with the OLS results.

```python
# -- WLS Regression --

# Weights: inverse of squared residual std dev
weights = 1 / (reg_clean['sigma'] ** 2)

X_wls = sm.add_constant(reg_clean[['cash_dummy', 'ln_deal_value', 'contested_dummy',
                                    'target_beta', 'target_sigma']])
model_wls = sm.WLS(reg_clean['CAR_pct'], X_wls, weights=weights).fit(cov_type='HC1')

# Comparison: OLS vs WLS
comparison_ols_wls = summary_col([model3, model_wls],
                                  model_names=['OLS (HC1)', 'WLS (HC1)'],
                                  stars=True,
                                  info_dict={'N': lambda x: f'{int(x.nobs)}',
                                             'R-squared': lambda x: f'{x.rsquared:.3f}'})
print("OLS vs. WLS: Full Specification")
print("=" * 55)
print(comparison_ols_wls)
```

## 9. Regression Diagnostics

Several diagnostic checks are standard for cross-sectional regressions in event studies.

**Residual analysis.** The OLS residuals $\hat{\eta}_i$ should not show obvious patterns when plotted against fitted values or explanatory variables. A funnel shape (residuals fanning out) indicates heteroskedasticity; clustering indicates omitted non-linearities or interactions.

**Influence diagnostics.** With small samples, a single observation can drive the regression results. Cook's distance $D_i$ measures the influence of observation $i$ on the full set of coefficient estimates. Observations with $D_i > 4/N$ are conventionally flagged as influential.

**Multicollinearity.** The variance inflation factor for each regressor indicates whether collinearity inflates the standard errors. $VIF_k > 10$ is the conventional threshold for concern.

**Normality of residuals.** While OLS is consistent regardless of the residual distribution, the t-tests on individual coefficients rely on approximate normality of $\hat{\gamma}_k$, which requires either normal residuals or a large sample. With $N = 10$, this approximation is fragile, and the Jarque-Bera test provides a useful diagnostic.

```python
# -- Regression Diagnostics --

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

resid = model3.resid
fitted = model3.fittedvalues

# Panel A: Residuals vs. fitted values
ax = axes[0, 0]
ax.scatter(fitted, resid, s=80, color='#1f78b4', zorder=3)
ax.axhline(0, color='red', linestyle='--', linewidth=1)
for i, r in reg_clean.iterrows():
    ax.annotate(r['target'], (fitted.loc[i], resid.loc[i]), fontsize=7, ha='left')
ax.set_xlabel('Fitted CAR (%)')
ax.set_ylabel('Residual (%)')
ax.set_title('Panel A: Residuals vs. Fitted Values')
ax.grid(True, alpha=0.2)

# Panel B: Q-Q plot of residuals
ax = axes[0, 1]
(osm, osr), (slope, intercept, r_val) = stats.probplot(resid, dist='norm')
ax.scatter(osm, osr, s=60, color='#1f78b4', zorder=3)
ax.plot(osm, slope * osm + intercept, 'r-', linewidth=1.5)
ax.set_xlabel('Theoretical Quantiles')
ax.set_ylabel('Sample Quantiles')
ax.set_title('Panel B: Q-Q Plot of Residuals')
ax.grid(True, alpha=0.2)

# Panel C: Cook's distance
ax = axes[1, 0]
influence = model3.get_influence()
cooks_d = influence.cooks_distance[0]
ax.bar(range(len(cooks_d)), cooks_d, color='#1f78b4', edgecolor='white', width=0.7)
ax.axhline(4 / len(cooks_d), color='red', linestyle='--', linewidth=1.5,
           label=f'Threshold: 4/N = {4/len(cooks_d):.2f}')
ax.set_xticks(range(len(cooks_d)))
ax.set_xticklabels(reg_clean['target'].values, rotation=45, fontsize=8)
ax.set_ylabel("Cook's Distance")
ax.set_title("Panel C: Cook's Distance")
ax.legend(fontsize=9, frameon=True)
ax.grid(True, alpha=0.2)

# Panel D: VIF
ax = axes[1, 1]
from statsmodels.stats.outliers_influence import variance_inflation_factor
X_vif = reg_clean[['cash_dummy', 'ln_deal_value', 'contested_dummy', 'target_beta', 'target_sigma']].values
X_vif_c = np.column_stack([np.ones(len(X_vif)), X_vif])
vif_names = ['const', 'Cash', 'ln(Deal Val)', 'Contested', 'Beta', 'Sigma']
vifs = [variance_inflation_factor(X_vif_c, i) for i in range(X_vif_c.shape[1])]
# Plot VIF for non-constant variables
ax.barh(vif_names[1:], vifs[1:], color='#1f78b4', edgecolor='white')
ax.axvline(5, color='orange', linestyle=':', linewidth=1.5, label='VIF = 5')
ax.axvline(10, color='red', linestyle='--', linewidth=1.5, label='VIF = 10')
ax.set_xlabel('VIF')
ax.set_title('Panel D: Variance Inflation Factors')
ax.legend(fontsize=9, frameon=True)
ax.grid(True, alpha=0.2)

plt.tight_layout()
plt.show()

# Jarque-Bera test on residuals
jb_stat, jb_p = stats.jarque_bera(resid)
print(f"Jarque-Bera test on OLS residuals: JB = {jb_stat:.2f}, p = {jb_p:.4f}")

influential = [(reg_clean.iloc[i]['target'], cooks_d[i]) for i in range(len(cooks_d)) if cooks_d[i] > 4/len(cooks_d)]
if influential:
    print(f"\nInfluential observations (Cook's D > {4/len(cooks_d):.2f}):")
    for name, d in influential:
        print(f"  {name}: D = {d:.3f}")
else:
    print("\nNo observations exceed the Cook's distance threshold.")
```

## 10. Robustness Checks

A credible cross-sectional event study should demonstrate that the results are not driven by specific modelling choices. Standard robustness checks include the following.

**Alternative event windows.** If the coefficient on $D_{cash}$ is positive and significant for $[-1, +1]$ but disappears for $[-5, +5]$, the result may be capturing pre-announcement dynamics rather than the announcement effect itself. Conversely, if the effect is stronger at $[-10, +10]$, there may be run-up associated with cash deals (e.g., because cash offers are more likely to leak).

**Dropping influential observations.** If Cook's distance identifies an observation with $D_i > 4/N$, re-estimating the regression without that observation tests whether the results are robust to the inclusion of an outlier.

**Alternative normal return models.** If the CARs are computed with a different normal return model (e.g., Fama-French three-factor rather than the market model), do the cross-sectional regression results change? This check is most relevant when one of the explanatory variables (e.g., target beta) is correlated with the choice of normal return model.

```python
# -- Robustness: alternative event windows --

robustness_rows = []
for tau_a, tau_b in [(-1, 1), (-5, 5), (-10, 10)]:
    car_col = f'CAR_[{tau_a:+d},{tau_b:+d}]'
    y_rob = reg_clean[car_col] * 100
    X_rob = sm.add_constant(reg_clean[['cash_dummy', 'ln_deal_value', 'contested_dummy']])
    m = sm.OLS(y_rob.dropna(), X_rob.loc[y_rob.dropna().index]).fit(cov_type='HC1')

    robustness_rows.append({
        'Window': f'[{tau_a:+d}, {tau_b:+d}]',
        'gamma_cash': m.params.get('cash_dummy', np.nan),
        'se_cash': m.bse.get('cash_dummy', np.nan),
        'p_cash': m.pvalues.get('cash_dummy', np.nan),
        'gamma_dealval': m.params.get('ln_deal_value', np.nan),
        'se_dealval': m.bse.get('ln_deal_value', np.nan),
        'p_dealval': m.pvalues.get('ln_deal_value', np.nan),
        'gamma_contest': m.params.get('contested_dummy', np.nan),
        'se_contest': m.bse.get('contested_dummy', np.nan),
        'p_contest': m.pvalues.get('contested_dummy', np.nan),
        'R2': m.rsquared,
        'N': int(m.nobs),
    })

rob_df = pd.DataFrame(robustness_rows)

print("Robustness: Key Coefficients Across Event Windows")
print("=" * 80)
print("(Coefficients in percentage points, HC1 standard errors in parentheses)")
print()
for _, r in rob_df.iterrows():
    stars_cash = '***' if r['p_cash'] < 0.01 else ('**' if r['p_cash'] < 0.05 else ('*' if r['p_cash'] < 0.10 else ''))
    stars_dv = '***' if r['p_dealval'] < 0.01 else ('**' if r['p_dealval'] < 0.05 else ('*' if r['p_dealval'] < 0.10 else ''))
    stars_ct = '***' if r['p_contest'] < 0.01 else ('**' if r['p_contest'] < 0.05 else ('*' if r['p_contest'] < 0.10 else ''))
    print(f"  Window {r['Window']}:  Cash = {r['gamma_cash']:+.2f}{stars_cash} ({r['se_cash']:.2f}), "
          f"ln(DealVal) = {r['gamma_dealval']:+.2f}{stars_dv} ({r['se_dealval']:.2f}), "
          f"Contested = {r['gamma_contest']:+.2f}{stars_ct} ({r['se_contest']:.2f}), "
          f"R2 = {r['R2']:.3f}")
```

## 11. Connection to the Earnings Response Coefficient

The cross-sectional regression in M&A studies has a direct parallel in the accounting literature: the **Earnings Response Coefficient (ERC)** regression. In an earnings announcement event study, the CAR is regressed on the earnings surprise:

$$
CAR_i = \gamma_0 + \gamma_1 \cdot SUE_i + \eta_i
$$

where $SUE_i$ is the standardized unexpected earnings (actual earnings minus analyst consensus, divided by the standard deviation of past forecast errors). The coefficient $\gamma_1$ is the ERC: it measures how many percentage points of abnormal return are associated with a one-standard-deviation earnings surprise.

The ERC has been studied extensively since Ball and Brown (1968). Its magnitude varies with firm characteristics: it is larger for firms with more persistent earnings (Collins and Kothari, 1989), higher growth options, and lower leverage. The ERC regression is thus a cross-sectional event study of exactly the same form as Equation (1), but applied to a different event type.

This parallel underlines a general point: the cross-sectional regression framework is not specific to M&A. It applies to any event study where firm-level CARs vary and where observable characteristics may explain the variation. Dividend announcements, CEO turnovers, regulatory changes, patent grants, and litigation outcomes have all been analyzed using this two-step procedure.

## 12. Endogeneity and Limitations of Cross-Sectional Regressions

The cross-sectional regression in Equation (1) is a reduced-form model, and the coefficients should be interpreted as conditional correlations rather than causal effects unless additional identification arguments are provided.

The most common endogeneity concern is **omitted variable bias**. If a firm characteristic $Z_i$ is correlated with both the CAR and one of the included regressors $X_{k,i}$, the coefficient $\hat{\gamma}_k$ captures both the direct effect of $X_k$ and the indirect effect of $Z$ operating through $X_k$. For example, if cash offers are more common for smaller targets, and smaller targets earn higher CARs for reasons unrelated to payment method (e.g., liquidity premium), the coefficient on $D_{cash}$ will be upward-biased.

**Self-selection** is a related concern. The payment method is not randomly assigned: it is chosen endogenously by the acquirer based on factors (information asymmetry, tax status, capital structure) that may also affect the target's valuation. The ideal experiment would randomly assign cash versus stock payment to a set of identical deals, but this is not feasible. The econometric solutions include instrumental variables (if a valid instrument is available) and Heckman-style selection models. In practice, most M&A event studies acknowledge the endogeneity concern and interpret the payment method coefficient as descriptive rather than causal.

**Small sample limitations.** With $N = 10$ deals, the regression has very few degrees of freedom. The coefficient estimates are noisy, the t-statistics have low power, and the influence of individual observations is high. This is a pedagogical sample; a research-grade study would typically use 100 or more deals. The qualitative patterns (cash premium, deal-size effect, bidder-competition effect) are well-established in larger samples (Betton et al., 2008).
