Long-horizon event studies introduce fundamental challenges that do not arise at short horizons. The cumulative abnormal return and the buy-and-hold abnormal return, which are approximately equal over a three-day window, diverge substantially over twelve or thirty-six months. The choice of normal return model, which matters little over three days, becomes the dominant source of uncertainty over long horizons. Test statistics that are well-specified at short horizons become severely mis-sized at long horizons due to cross-sectional dependence, skewness, and the compounding of small model misspecifications. The result is that long-horizon event studies are far more fragile than their short-horizon counterparts, and the literature has developed a specialized set of methods to address these problems.

## 1. Why Long-Horizon Event Studies?

Short-horizon event studies measure the market's immediate reaction to an event. Under the efficient market hypothesis, this reaction is the market's best estimate of the event's value impact, and no further drift should occur after the announcement. If markets are efficient, the short-horizon CAR is a sufficient statistic for the event's wealth effect, and long-horizon studies are unnecessary.

Long-horizon event studies exist precisely because this prediction sometimes fails. The empirical literature has documented several instances of post-event drift: prices continue to move in the direction of the initial reaction for months or years after the event. The three best-known examples are:

Post-earnings announcement drift (PEAD): stocks with positive earnings surprises continue to outperform for 60 to 90 trading days after the announcement (Bernard and Thomas, 1989). This was discussed briefly in Session 1.

IPO underperformance: firms that go public underperform matched firms by 20 to 30 percent over the three years following the offering (Ritter, 1991; Loughran and Ritter, 1995).

Post-merger acquirer underperformance: acquiring firms, especially those using stock as payment, underperform by 10 to 20 percent over the three years following the merger (Agrawal, Jaffe, and Mandelker, 1992; Loughran and Vijh, 1997).

These findings, if genuine, are direct challenges to the efficient market hypothesis. Fama (1998) argued that many of these "anomalies" are methodological artifacts: they are sensitive to the choice of normal return model, the measure of abnormal return (CAR vs. BHAR), and the test statistic. This methodological debate motivates the careful treatment of long-horizon methods in this session.

## 2. Cumulative Abnormal Returns vs. Buy-and-Hold Abnormal Returns

At short horizons, the CAR and the BHAR are nearly identical. At long horizons, they diverge because of the compounding effect. We now derive each measure and show precisely where and why they differ.

### 2.1 The Long-Horizon CAR

The CAR over a horizon of $T$ months (or trading days) is the sum of monthly abnormal returns:

$$
CAR_i(1, T) = \sum_{t=1}^{T} AR_{i,t} = \sum_{t=1}^{T} \left(R_{i,t} - E[R_{i,t}]\right) \tag{1}
$$

This measure implicitly assumes monthly rebalancing: at the end of each month, the investor resets the position to one dollar in the event firm and one dollar short in the benchmark. This is a "paper" return that no investor actually earns.

### 2.2 The Buy-and-Hold Abnormal Return

The BHAR measures the return from buying the event firm on the event date and holding it for $T$ periods, minus the return from a similar buy-and-hold strategy in the benchmark:

$$
BHAR_i(1, T) = \prod_{t=1}^{T}(1 + R_{i,t}) - \prod_{t=1}^{T}(1 + E[R_{i,t}]) \tag{2}
$$

This measure corresponds to the actual wealth difference an investor would experience: it is the dollar difference between holding one dollar in the event firm and one dollar in the benchmark over the full horizon.

### 2.3 The Divergence

The difference between CAR and BHAR is a cross-product term that grows with the horizon:

$$
BHAR_i \approx CAR_i + \text{compounding term}
$$

To see this, consider a two-period example. Let $R_1 = 0.10$ and $R_2 = 0.10$ for the event firm, and $E[R_1] = E[R_2] = 0.05$ for the benchmark. Then:

$$
CAR = (0.10 - 0.05) + (0.10 - 0.05) = 0.10
$$

$$
BHAR = (1.10)(1.10) - (1.05)(1.05) = 1.21 - 1.1025 = 0.1075
$$

The BHAR exceeds the CAR by 0.0075, which is the cross-product of the abnormal returns. Over 36 months, these cross-products cumulate and the discrepancy between CAR and BHAR can be large, especially when abnormal returns are themselves large (as in the IPO and M&A settings).

Barber and Lyon (1997) argued that the BHAR is the economically relevant measure because it captures the investor's actual experience. The CAR overstates negative long-run performance and understates positive long-run performance relative to the BHAR, because it ignores compounding. Fama (1998) countered that the BHAR is more susceptible to the "bad model" problem: because it compounds misspecification of the normal return model, even a small monthly bias in the benchmark (say, 0.1% per month) accumulates to 3.6% over 36 months. The CAR, by summing rather than compounding, is more robust to small misspecifications.

## 3. The Bad Model Problem

The core difficulty of long-horizon event studies is that *every* model of expected returns is misspecified to some degree. At short horizons, this misspecification is negligible: if the true expected daily return differs from the model's prediction by 0.01%, this amounts to 0.03% over a three-day window, which is tiny relative to the signal. At long horizons, the same 0.01% daily misspecification compounds to approximately 2.5% per year, or 7.5% over three years, which is of the same order of magnitude as the abnormal returns being measured.

Kothari and Warner (1997) formalized this problem. They showed that when the normal return model is misspecified, the CAR and BHAR are biased, and the bias grows linearly (for the CAR) or exponentially (for the BHAR) with the horizon. The direction of the bias depends on the sign of the misspecification:

If the model overstates expected returns (e.g., because it assigns too high a beta to event firms), the measured abnormal return will be systematically negative, generating spurious underperformance.

If the model understates expected returns, the measured abnormal return will be systematically positive.

This is why long-horizon event studies are so sensitive to the choice of benchmark. The market model, which is adequate at short horizons, is generally insufficient at long horizons because it does not account for the size, value, and momentum effects that drive cross-sectional return differences over extended periods. The Fama-French three-factor model improves the benchmark, but it too is misspecified to some degree.

The reference-portfolio approach (matching event firms to control firms of similar size and book-to-market) and the calendar-time portfolio approach (Section 7) are both attempts to mitigate the bad model problem.

## 4. Application: Post-Merger Acquirer Performance

We study the long-run stock price performance of acquiring firms following completed mergers. This is the canonical application in the long-horizon event study literature. The sample consists of large U.S. acquisitions announced between 2018 and 2021, with the "event date" defined as the first trading day of the month following deal completion. We track acquirer returns for 12 months after completion.

The sample is deliberately kept small for pedagogical clarity. In a research setting, one would use the full universe of deals from SDC Platinum and track returns for 36 or 60 months.

```python
import pandas as pd
import numpy as np
import yfinance as yf
import statsmodels.api as sm
import matplotlib.pyplot as plt
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# -- Sample: acquirers after deal completion --
acquirers = pd.DataFrame({
    'ticker': ['MSFT', 'AVGO', 'ABBV', 'LOW', 'TXN', 'BDX', 'DIS', 'ORCL', 'COP', 'UBER'],
    'target_name': ['Nuance', 'CA Technologies', 'Allergan', "?"+'s National Express",
                     'Maxim Integrated', 'C.R. Bard', '21st Century Fox', 'Cerner',
                     'Concho Resources', 'Postmates'],
    'completion_date': pd.to_datetime([
        '2022-03-04', '2018-11-05', '2020-05-08', '2022-01-03',
        '2021-08-26', '2019-01-01', '2019-03-20', '2022-06-08',
        '2021-01-15', '2020-12-01'
    ]),
    'deal_value_bn': [19.7, 18.9, 63.0, 7.4, 27.0, 24.0, 71.3, 28.3, 9.7, 2.65],
    'payment': ['Cash', 'Cash', 'Cash+Stock', 'Cash', 'Cash+Stock',
                'Cash+Stock', 'Cash+Stock', 'Cash+Stock', 'Stock', 'Stock'],
})

print("Acquirer Sample: Post-Merger Long-Run Performance")
print("=" * 80)
print(acquirers[['ticker', 'target_name', 'completion_date', 'deal_value_bn', 'payment']].to_string(index=False))
```

```python
# -- Download monthly returns --
start_dl = '2016-01-01'
end_dl = '2024-06-01'
market_ticker = '^GSPC'

data_mkt = yf.download(market_ticker, start=start_dl, end=end_dl, interval='1mo', progress=False)
if isinstance(data_mkt.columns, pd.MultiIndex):
    data_mkt.columns = data_mkt.columns.get_level_values(0)
mkt_monthly = data_mkt['Close'].pct_change().dropna()

acq_tickers = acquirers['ticker'].tolist()
data_acq = yf.download(acq_tickers, start=start_dl, end=end_dl, interval='1mo', progress=False)
if isinstance(data_acq.columns, pd.MultiIndex):
    acq_prices_m = data_acq['Close']
else:
    acq_prices_m = data_acq[['Close']].rename(columns={'Close': acq_tickers[0]})

acq_monthly = acq_prices_m.pct_change().dropna()
common_months = mkt_monthly.index.intersection(acq_monthly.index)
mkt_monthly = mkt_monthly.loc[common_months]
acq_monthly = acq_monthly.loc[common_months]

print(f"Monthly data: {len(common_months)} months ({common_months[0].date()} to {common_months[-1].date()})")
print(f"Acquirers downloaded: {acq_monthly.shape[1]}")
```

## 5. Computing Long-Horizon CARs and BHARs

We compute both measures for each acquirer over horizons of 3, 6, and 12 months following deal completion. The normal return benchmark is the market return (S&P 500). In a research setting, one would also use characteristic-matched portfolios (size and book-to-market) or the Fama-French three-factor model as benchmarks.

For the CAR, we sum the monthly abnormal returns:

$$
CAR_i(1, T) = \sum_{t=1}^{T} (R_{i,t} - R_{m,t})
$$

For the BHAR, we compound:

$$
BHAR_i(1, T) = \prod_{t=1}^{T}(1 + R_{i,t}) - \prod_{t=1}^{T}(1 + R_{m,t})
$$

We use the market-adjusted return (subtracting the market return directly) rather than the market model residual because estimating a monthly beta with precision requires a long estimation window, and we want to keep the procedure transparent.

```python
# -- Compute CARs and BHARs --
horizons = [3, 6, 12]
results_list = []

for _, acq in acquirers.iterrows():
    tick = acq['ticker']
    comp_date = acq['completion_date']

    if tick not in acq_monthly.columns:
        continue

    # Find the first month after completion
    post_months = acq_monthly.index[acq_monthly.index >= comp_date]
    if len(post_months) == 0:
        continue

    row = {
        'ticker': tick,
        'target': acq['target_name'],
        'completion': comp_date,
        'deal_value_bn': acq['deal_value_bn'],
        'payment': acq['payment'],
    }

    for T in horizons:
        if len(post_months) < T:
            row[f'CAR_{T}m'] = np.nan
            row[f'BHAR_{T}m'] = np.nan
            continue

        r_firm = acq_monthly.loc[post_months[:T], tick].values
        r_mkt = mkt_monthly.loc[post_months[:T]].values

        # Handle NaNs
        valid = ~(np.isnan(r_firm) | np.isnan(r_mkt))
        r_firm_c = r_firm[valid]
        r_mkt_c = r_mkt[valid]

        if len(r_firm_c) < T * 0.5:
            row[f'CAR_{T}m'] = np.nan
            row[f'BHAR_{T}m'] = np.nan
            continue

        # CAR: sum of monthly abnormal returns
        car = np.sum(r_firm_c - r_mkt_c)

        # BHAR: compound difference
        bhar = np.prod(1 + r_firm_c) - np.prod(1 + r_mkt_c)

        row[f'CAR_{T}m'] = car
        row[f'BHAR_{T}m'] = bhar

    results_list.append(row)

lh_df = pd.DataFrame(results_list)

print("Long-Horizon Abnormal Returns (%)")
print("=" * 90)
display_df = lh_df[['ticker', 'target', 'payment']].copy()
for T in horizons:
    display_df[f'CAR {T}m'] = lh_df[f'CAR_{T}m'] * 100
    display_df[f'BHAR {T}m'] = lh_df[f'BHAR_{T}m'] * 100
print(display_df.to_string(index=False, float_format='{:.2f}'.format))

print(f"\nCross-sectional averages:")
for T in horizons:
    car_mean = lh_df[f'CAR_{T}m'].mean() * 100
    bhar_mean = lh_df[f'BHAR_{T}m'].mean() * 100
    print(f"  {T} months: CAAR = {car_mean:+.2f}%,  Avg BHAR = {bhar_mean:+.2f}%,  "
          f"Diff = {bhar_mean - car_mean:+.2f}%")
```

## 6. Visualizing CAR vs. BHAR Divergence

The numerical comparison above shows the aggregate divergence at three horizons. The plots below trace the month-by-month evolution of the CAR and BHAR for each acquirer and for the cross-sectional average, making the compounding effect visible.

```python
# -- Month-by-month CAR and BHAR paths --

fig, axes = plt.subplots(1, 2, figsize=(14, 5.5))
max_T = 12

car_paths = {}
bhar_paths = {}

for _, acq in acquirers.iterrows():
    tick = acq['ticker']
    comp_date = acq['completion_date']
    if tick not in acq_monthly.columns:
        continue

    post_months = acq_monthly.index[acq_monthly.index >= comp_date]
    if len(post_months) < max_T:
        continue

    r_firm = acq_monthly.loc[post_months[:max_T], tick].values
    r_mkt = mkt_monthly.loc[post_months[:max_T]].values
    valid = ~(np.isnan(r_firm) | np.isnan(r_mkt))
    if valid.sum() < max_T * 0.5:
        continue

    ar = r_firm - r_mkt
    car_path = np.concatenate([[0], np.cumsum(np.where(valid, ar, 0))])
    bhar_path_firm = np.concatenate([[1], np.cumprod(np.where(valid, 1 + r_firm, 1))])
    bhar_path_mkt = np.concatenate([[1], np.cumprod(np.where(valid, 1 + r_mkt, 1))])
    bhar_path = bhar_path_firm - bhar_path_mkt

    car_paths[tick] = car_path
    bhar_paths[tick] = bhar_path

months = np.arange(0, max_T + 1)

# Panel A: Individual CAR and BHAR paths
ax = axes[0]
for tick in car_paths:
    ax.plot(months, car_paths[tick] * 100, '-', alpha=0.4, linewidth=1)
# Average
if car_paths:
    avg_car = np.mean([car_paths[t] for t in car_paths], axis=0)
    ax.plot(months, avg_car * 100, 'k-', linewidth=2.5, label='Average CAR')
ax.axhline(0, color='black', linewidth=0.5)
ax.set_xlabel('Months After Completion')
ax.set_ylabel('CAR (%)')
ax.set_title('Panel A: Cumulative Abnormal Returns')
ax.legend(fontsize=9, frameon=True)
ax.grid(True, alpha=0.2)

# Panel B: Average CAR vs Average BHAR
ax = axes[1]
if car_paths and bhar_paths:
    avg_car = np.mean([car_paths[t] for t in car_paths], axis=0)
    avg_bhar = np.mean([bhar_paths[t] for t in bhar_paths], axis=0)
    ax.plot(months, avg_car * 100, 'o-', color='#1f78b4', linewidth=2, markersize=4, label='Average CAR')
    ax.plot(months, avg_bhar * 100, 's-', color='#e31a1c', linewidth=2, markersize=4, label='Average BHAR')
    ax.fill_between(months, avg_car * 100, avg_bhar * 100, alpha=0.15, color='gray', label='Compounding wedge')
ax.axhline(0, color='black', linewidth=0.5)
ax.set_xlabel('Months After Completion')
ax.set_ylabel('Abnormal Return (%)')
ax.set_title('Panel B: CAR vs. BHAR (Cross-Sectional Average)')
ax.legend(fontsize=9, frameon=True)
ax.grid(True, alpha=0.2)

plt.tight_layout()
plt.show()
```

## 7. Testing Long-Horizon Abnormal Returns

The standard event-time test for long-horizon abnormal returns is the cross-sectional t-test (the same Brown-Warner test from Session 4), applied to the BHARs or CARs:

$$
t = \frac{\overline{BHAR}}{\hat{S}_{BHAR} / \sqrt{N}}
$$

However, this test has well-documented problems at long horizons (Barber and Lyon, 1997; Lyon, Barber, and Tsai, 1999):

**Positive skewness of BHARs.** Because the BHAR involves compounding, its distribution is right-skewed. The t-test, which is based on the mean, is sensitive to this skewness: a few firms with very large positive BHARs can pull the average upward, causing over-rejection of the null on the positive side and under-rejection on the negative side.

**Cross-sectional dependence.** When event firms have overlapping post-event windows (which is almost always the case in long-horizon studies, since events span months or years), their BHARs share common market-wide shocks. The cross-sectional variance underestimates the true variance, leading to over-rejection. This is the same problem as calendar-time clustering (Session 4), but it is far more severe at long horizons because the windows are longer and overlap more.

**New listing bias.** The benchmark portfolio's composition changes over the post-event horizon as new firms enter and old firms exit. If new listings systematically underperform (as documented by Ritter, 1991), the benchmark return is biased downward, inflating the measured abnormal return.

We compute the simple cross-sectional t-test below, noting its limitations. Section 8 introduces the calendar-time portfolio approach, which addresses the cross-sectional dependence problem directly.

```python
# -- Event-Time Tests --

print("Event-Time Tests of Long-Horizon Abnormal Returns")
print("=" * 70)

for T in horizons:
    car_col = f'CAR_{T}m'
    bhar_col = f'BHAR_{T}m'

    cars = lh_df[car_col].dropna()
    bhars = lh_df[bhar_col].dropna()
    N = len(bhars)

    if N < 3:
        print(f"  {T} months: insufficient observations (N={N})")
        continue

    # CAR test
    car_mean = cars.mean()
    car_se = cars.std(ddof=1) / np.sqrt(N)
    car_t = car_mean / car_se
    car_p = 2 * (1 - stats.t.cdf(abs(car_t), df=N - 1))

    # BHAR test
    bhar_mean = bhars.mean()
    bhar_se = bhars.std(ddof=1) / np.sqrt(N)
    bhar_t = bhar_mean / bhar_se
    bhar_p = 2 * (1 - stats.t.cdf(abs(bhar_t), df=N - 1))

    # Skewness-adjusted t-test (Lyon, Barber, Tsai 1999)
    skew_bhar = stats.skew(bhars)
    sa_t = bhar_t + (1/3) * skew_bhar * (1/np.sqrt(N)) + (1/6) * skew_bhar**2 * (1/N)

    print(f"\n  Horizon: {T} months (N = {N})")
    print(f"    CAR:  mean = {car_mean*100:+.2f}%, t = {car_t:.3f}, p = {car_p:.4f}")
    print(f"    BHAR: mean = {bhar_mean*100:+.2f}%, t = {bhar_t:.3f}, p = {bhar_p:.4f}")
    print(f"    BHAR skewness = {skew_bhar:.3f}, skewness-adjusted t = {sa_t:.3f}")
```

## 8. The Calendar-Time Portfolio Approach

The calendar-time portfolio approach, proposed by Jaffe (1974) and Mandelker (1974) and advocated by Fama (1998) and Mitchell and Stafford (2000), addresses the cross-sectional dependence problem by aggregating in calendar time rather than event time.

The procedure is as follows. In each calendar month $t$, form a portfolio consisting of all firms that are within the post-event window (e.g., within 12 months of their deal completion). The portfolio return in month $t$ is:

$$
R_{p,t} = \frac{1}{N_t} \sum_{i \in S_t} R_{i,t}
$$

where $S_t$ is the set of firms in the portfolio in month $t$ and $N_t$ is the number of firms. This is an equal-weighted calendar-time portfolio. The portfolio composition changes each month as firms enter (new deal completions) and exit (reaching the end of the post-event window).

The abnormal return of the portfolio is estimated by regressing $R_{p,t}$ on the market factor (or a multi-factor model):

$$
R_{p,t} - R_{f,t} = \alpha + \beta (R_{m,t} - R_{f,t}) + \varepsilon_t \tag{3}
$$

The intercept $\alpha$ is the average monthly abnormal return of the portfolio of event firms. The t-statistic on $\alpha$ provides a test of the null hypothesis that event firms earn zero abnormal returns. This t-statistic is computed from the time series of portfolio returns and therefore automatically accounts for any cross-sectional correlation: if event firms share common shocks, these shocks affect the portfolio return on the same calendar date and are captured in the regression residual $\varepsilon_t$.

The main advantages of the calendar-time approach are: (a) it eliminates the cross-sectional dependence problem entirely, because the unit of observation is a portfolio return, not a firm-level return; (b) it uses standard time-series regression, for which the econometrics are well understood; and (c) it can easily accommodate multi-factor benchmarks (CAPM, Fama-French 3-factor, Carhart 4-factor).

The main disadvantage is that it pools together firms at different points in their post-event history: in any given month, the portfolio contains firms that completed their deal one month ago alongside firms that completed 11 months ago. This averaging across event-time positions reduces power to detect abnormal performance that is concentrated at a specific point in the post-event period (e.g., in the first three months).

```python
# -- Calendar-Time Portfolio --

holding_period = 12  # months

# Build calendar-time portfolio
portfolio_returns = {}

for month in acq_monthly.index:
    # Firms in the portfolio this month: completed deal within last 12 months
    in_portfolio = []
    for _, acq in acquirers.iterrows():
        tick = acq['ticker']
        comp = acq['completion_date']
        if tick not in acq_monthly.columns:
            continue
        months_since = (month.year - comp.year) * 12 + (month.month - comp.month)
        if 0 < months_since <= holding_period:
            ret = acq_monthly.loc[month, tick] if month in acq_monthly.index else np.nan
            if not np.isnan(ret):
                in_portfolio.append(ret)

    if len(in_portfolio) >= 1:
        portfolio_returns[month] = {
            'R_p': np.mean(in_portfolio),
            'N': len(in_portfolio),
        }

port_df = pd.DataFrame(portfolio_returns).T
port_df.index = pd.DatetimeIndex(port_df.index)
port_df = port_df.sort_index()

# Merge with market returns
port_df['R_m'] = mkt_monthly.reindex(port_df.index)
port_df = port_df.dropna()

print(f"Calendar-time portfolio: {len(port_df)} months with at least 1 firm")
print(f"Average number of firms per month: {port_df['N'].mean():.1f}")
print(f"Range: {port_df.index[0].date()} to {port_df.index[-1].date()}")
```

```python
# -- CAPM regression on calendar-time portfolio --

# Use risk-free rate = 0 for simplicity (monthly RF is small)
y = port_df['R_p'] * 100  # in pct
X = sm.add_constant(port_df['R_m'] * 100)

capm_reg = sm.OLS(y, X).fit(cov_type='HAC', cov_kwds={'maxlags': 3})

print("Calendar-Time Portfolio: CAPM Regression")
print("=" * 60)
print(f"  Alpha (monthly): {capm_reg.params['const']:+.3f}% (t = {capm_reg.tvalues['const']:.3f}, "
      f"p = {capm_reg.pvalues['const']:.4f})")
print(f"  Beta:            {capm_reg.params['R_m']:.3f} (t = {capm_reg.tvalues['R_m']:.3f})")
print(f"  R-squared:       {capm_reg.rsquared:.3f}")
print(f"  Months:          {int(capm_reg.nobs)}")
print(f"  Alpha annualized: {capm_reg.params['const'] * 12:+.2f}%")
print(f"\n  Standard errors: Newey-West HAC (3 lags)")

# Plot
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

ax = axes[0]
ax.bar(port_df.index, (port_df['R_p'] - port_df['R_m']) * 100,
       width=25, color='#1f78b4', edgecolor='white', alpha=0.7)
ax.axhline(capm_reg.params['const'], color='red', linestyle='--', linewidth=1.5,
           label=f'Alpha = {capm_reg.params["const"]:+.3f}%/month')
ax.axhline(0, color='black', linewidth=0.5)
ax.set_ylabel('Portfolio Excess Return over Market (%)')
ax.set_title('Calendar-Time Portfolio: Monthly Abnormal Returns')
ax.legend(fontsize=9, frameon=True)
ax.grid(True, alpha=0.2)

ax = axes[1]
ax2 = ax.twinx()
ax.plot(port_df.index, np.cumsum((port_df['R_p'] - port_df['R_m'])) * 100,
        '-', color='#1f78b4', linewidth=2, label='Cumulative Abnormal Return')
ax2.bar(port_df.index, port_df['N'], width=25, alpha=0.3, color='gray', label='N firms')
ax.axhline(0, color='black', linewidth=0.5)
ax.set_ylabel('Cumulative Abnormal Return (%)', color='#1f78b4')
ax2.set_ylabel('Number of Firms in Portfolio', color='gray')
ax.set_title('Calendar-Time Portfolio: Cumulative AR and Composition')
lines1, labels1 = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax.legend(lines1 + lines2, labels1 + labels2, fontsize=9, frameon=True, loc='upper left')
ax.grid(True, alpha=0.2)

plt.tight_layout()
plt.show()
```

## 9. Comparing Event-Time and Calendar-Time Results

We now compare the three approaches to measuring long-run abnormal performance:

The event-time CAR provides one estimate: the cumulative sum of monthly abnormal returns, averaged across firms. Its t-test does not account for cross-sectional dependence.

The event-time BHAR provides a second estimate: the compounded abnormal return, which reflects the investor's actual wealth experience. Its t-test suffers from both skewness bias and cross-sectional dependence.

The calendar-time portfolio alpha provides a third estimate: the average monthly abnormal return of the portfolio of event firms, estimated from a time-series regression that automatically accounts for cross-sectional dependence.

When these three estimates agree in sign and significance, the conclusion is robust. When they disagree, the pattern is informative. If the event-time tests reject but the calendar-time alpha does not, the event-time tests are likely over-rejecting due to cross-sectional dependence. If the BHAR is much larger than the CAR, compounding effects are important and the researcher should report the BHAR as the economically relevant measure while acknowledging its statistical fragility.

```python
# -- Summary comparison table --

print("Comparison of Long-Run Abnormal Return Estimates")
print("=" * 75)
print(f"\nEvent-Time Estimates (cross-sectional average, {len(lh_df)} firms):")
for T in [3, 6, 12]:
    car_col = f'CAR_{T}m'
    bhar_col = f'BHAR_{T}m'
    cars = lh_df[car_col].dropna()
    bhars = lh_df[bhar_col].dropna()
    N = len(bhars)
    if N >= 2:
        car_t = cars.mean() / (cars.std(ddof=1) / np.sqrt(N))
        bhar_t = bhars.mean() / (bhars.std(ddof=1) / np.sqrt(N))
        print(f"  {T:2d} months:  CAR = {cars.mean()*100:+6.2f}% (t={car_t:+.2f}), "
              f"BHAR = {bhars.mean()*100:+6.2f}% (t={bhar_t:+.2f})")

print(f"\nCalendar-Time Portfolio (CAPM, {int(capm_reg.nobs)} months):")
print(f"  Alpha = {capm_reg.params['const']:+.3f}%/month "
      f"(t = {capm_reg.tvalues['const']:+.3f}, "
      f"annualized = {capm_reg.params['const']*12:+.2f}%)")

print(f"\nInterpretation:")
alpha_sign = "positive" if capm_reg.params['const'] > 0 else "negative"
alpha_sig = "significant" if capm_reg.pvalues['const'] < 0.10 else "insignificant"
print(f"  Calendar-time alpha is {alpha_sign} and {alpha_sig} at 10%.")
```

## 10. The Methodological Debate

The choice between event-time and calendar-time methods has been the subject of a sustained debate in the literature. The key positions are as follows.

**Fama (1998)** argued that long-horizon anomalies are largely methodological artifacts. He showed that most findings of long-run abnormal performance are sensitive to whether the researcher uses CARs or BHARs, equally-weighted or value-weighted portfolios, and the market model or a multi-factor benchmark. He advocated the calendar-time portfolio approach as the most reliable method, because it avoids the cross-sectional dependence problem and uses standard time-series inference.

**Loughran and Ritter (2000)** countered that the calendar-time approach has lower power to detect abnormal performance than the event-time BHAR approach. The reason is that the calendar-time portfolio weights each calendar month equally, not each event equally. In months when few events occur, the portfolio return is noisy; in months when many events occur, the portfolio return is precise. The equal weighting of months dilutes the signal. Loughran and Ritter argued that the event-time BHAR, despite its statistical problems, is more powerful and more economically meaningful.

**Mitchell and Stafford (2000)** showed that when the cross-sectional dependence is properly accounted for (by estimating the covariance matrix of BHARs), the significance of many long-horizon anomalies disappears. They argued that the standard event-time tests are over-sized by a factor of two or more when post-event windows overlap extensively.

The current consensus is that a well-executed long-horizon event study should report both event-time and calendar-time results. If both approaches yield the same conclusion, the finding is credible. If they disagree, the researcher should favor the calendar-time results for inference (because of better size control) while reporting the event-time BHARs for economic interpretation (because of their direct wealth relevance).

## 11. Practical Recommendations

The following guidelines distill the methodological literature into actionable advice for applied researchers.

**Report both CAR and BHAR.** The CAR is more robust to the bad-model problem; the BHAR is more economically meaningful. Report both and discuss the divergence.

**Use the calendar-time portfolio for inference.** The t-statistic on the calendar-time alpha is the most reliable test of the null hypothesis of zero abnormal returns, because it accounts for cross-sectional dependence. Use Newey-West standard errors to correct for any autocorrelation in the portfolio returns.

**Use a multi-factor benchmark.** At horizons beyond a few months, the single-factor market model is inadequate. The Fama-French three-factor or Carhart four-factor model substantially reduces the bad-model problem. If the abnormal return is robust across benchmarks, the finding is more credible.

**Match on characteristics.** For the event-time BHAR, use control firms matched on size and book-to-market ratio (or size, book-to-market, and momentum) as the benchmark, rather than a broad market index. This reduces the bad-model bias because the benchmark has the same factor exposures as the event firm.

**Be cautious about horizons beyond 36 months.** The statistical and economic problems compound with the horizon. At five years, the bad-model problem, cross-sectional dependence, and survivorship bias are all severe, and the results are almost never robust across methods.

**Report the cross-sectional distribution.** The average BHAR can mask substantial heterogeneity. Report the median BHAR (which is less sensitive to outliers than the mean) and the fraction of firms with negative BHARs.
