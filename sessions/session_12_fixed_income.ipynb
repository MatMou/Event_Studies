The entire course so far has used equity returns as the outcome variable. Equity markets are liquid, returns are available at daily or intraday frequency, and the market model provides a well-understood benchmark. But many of the most important events in finance, including sovereign credit downgrades, central bank policy announcements, corporate defaults, and regulatory interventions, have their primary impact on bond markets and credit spreads rather than on equity prices. An event study confined to equities misses these effects entirely, or captures them only indirectly through the equity-credit linkage.

This session extends the event study methodology to fixed income markets. The extension is not trivial. Bond markets differ from equity markets in ways that affect every step of the event study: the outcome variable is not a return but a yield or a spread; the distribution of yield changes is different from the distribution of equity returns; bond prices are often stale due to illiquidity; and the appropriate normal model depends on the term structure and credit curve, not on a single market factor.

We develop the methodology in three stages. First, we study events in sovereign bond markets using changes in government bond yields and sovereign CDS spreads. Second, we study events in corporate credit markets using changes in corporate bond spreads and CDS spreads. Third, we address the econometric challenges specific to fixed income event studies: illiquidity, stale prices, and the choice of benchmark.

The application is the impact of sovereign credit rating downgrades on government bond yields and CDS spreads, one of the most studied topics in sovereign finance.

## 1. Why Fixed Income Event Studies Differ from Equity Event Studies

### 1.1 The Outcome Variable

In equity event studies, the outcome is the stock return $R_{i,t} = (P_t - P_{t-1}) / P_{t-1}$, which is unit-free, approximately normally distributed at daily frequency, and comparable across firms. In fixed income markets, the natural outcome variable depends on the context.

For sovereign bonds, the standard outcome is the change in the government bond yield: $\Delta y_{i,t} = y_{i,t} - y_{i,t-1}$, measured in basis points. Yield changes are approximately normally distributed and stationary, making them well-suited for event study analysis. The yield level itself is not a useful outcome because it is non-stationary.

For credit-sensitive instruments, the outcome is the change in the credit spread: $\Delta s_{i,t} = s_{i,t} - s_{i,t-1}$, where the spread is measured relative to a risk-free benchmark (typically the government bond yield of the same maturity, or the swap rate). The spread isolates the credit component of the yield, removing the interest rate component.

For credit default swaps (CDS), the outcome is the change in the CDS spread $\Delta CDS_{i,t}$ in basis points. CDS spreads are available at daily frequency, are more liquid than many bond markets (especially for sovereign issuers and large corporates), and provide a direct measure of the market's assessment of default risk.

### 1.2 The Normal Model

The equity event study uses the market model as the benchmark: $E[R_{i,t}] = \alpha_i + \beta_i R_{m,t}$. The fixed income equivalent depends on the application.

For sovereign bond yields, the simplest benchmark is the yield change of a reference sovereign (e.g., the U.S. Treasury or the German Bund): $E[\Delta y_{i,t}] = \alpha_i + \beta_i \Delta y_{ref,t}$. This is analogous to the market model but applied to yield changes. The coefficient $\beta_i$ captures the sensitivity of country $i$'s yield to movements in the global risk-free rate.

For corporate bond spreads, the benchmark is the change in a matched spread index (e.g., the ICE BofA index for the same rating category and maturity bucket): $E[\Delta s_{i,t}] = \alpha_i + \beta_i \Delta s_{index,t}$. This controls for market-wide credit spread movements that are unrelated to the firm-specific event.

For CDS spreads, the same logic applies: the benchmark is a CDS index (e.g., the CDX North America Investment Grade index or the iTraxx Europe index).

### 1.3 Illiquidity

The single most important difference between equity and fixed income event studies is liquidity. Many bonds trade infrequently: corporate bonds may trade only a few times per day (or not at all), and sovereign bonds of smaller issuers may have wide bid-ask spreads and stale quotes. This creates two problems.

First, the observed yield or spread may not reflect the current market price but rather the price of the last trade, which may have occurred hours or even days earlier. This is the stale price problem, analogous to the thin trading problem in equity event studies (Session 2) but far more severe.

Second, the bid-ask bounce in illiquid bonds induces spurious negative autocorrelation in yield changes, inflating the variance of the abnormal yield change and reducing the power of the event study.

The CDS market partially mitigates these problems because CDS contracts are standardized and trade more frequently than the underlying bonds. For this reason, many fixed income event studies use CDS spreads rather than bond yields as the outcome variable, even when the research question concerns bonds.

## 2. Application: Sovereign Credit Rating Downgrades

### 2.1 Background

Sovereign credit ratings, assigned by agencies such as S&P, Moody's, and Fitch, summarize the agency's assessment of a government's ability and willingness to service its debt. A downgrade signals increased default risk and typically triggers a repricing of the sovereign's bonds and CDS.

The academic literature on sovereign rating events is extensive (Cantor and Packer, 1996; Afonso et al., 2012; Ismailescu and Kazemi, 2010). The key findings are: downgrades produce significant increases in sovereign bond yields and CDS spreads, especially for emerging market issuers; upgrades produce smaller and often insignificant decreases in spreads (an asymmetry attributed to the fact that upgrades are more anticipated); and there is substantial cross-border contagion, with a downgrade of one country increasing spreads in neighboring countries.

### 2.2 Sample

We study a sample of sovereign credit rating downgrades by S&P between 2010 and 2023. For each downgrade, we observe the country, the date, the old and new rating, and we measure the impact on the sovereign CDS spread. We use CDS data because it is more liquid than sovereign bond data for many of the affected issuers.

We proxy sovereign CDS dynamics using country ETFs and Treasury yields, since free daily CDS data is not readily available via `yfinance`. Specifically, we use changes in country-specific government bond ETF prices (which move inversely with yields) and changes in the spread between the country ETF yield and the U.S. Treasury yield as our outcome variables.

```python
import pandas as pd
import numpy as np
import yfinance as yf
import statsmodels.api as sm
import matplotlib.pyplot as plt
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# -- Sample: sovereign credit rating downgrades (S&P) --
downgrades = pd.DataFrame({
    'country': ['Italy', 'Spain', 'France', 'Brazil', 'Turkey',
                'South Africa', 'UK', 'China', 'Russia', 'Japan'],
    'date': pd.to_datetime([
        '2011-09-19',  # Italy AA- to A+
        '2012-01-13',  # Spain AA- to A
        '2012-01-13',  # France AAA to AA+
        '2015-09-09',  # Brazil BBB- to BB+
        '2016-07-20',  # Turkey BB+ to BB
        '2017-04-03',  # South Africa BBB- to BB+
        '2016-06-27',  # UK AAA to AA (post-Brexit)
        '2017-09-21',  # China AA- to A+
        '2014-04-25',  # Russia BBB to BBB-
        '2011-01-27',  # Japan AA to AA-
    ]),
    'old_rating': ['AA-', 'AA-', 'AAA', 'BBB-', 'BB+', 'BBB-', 'AAA', 'AA-', 'BBB', 'AA'],
    'new_rating': ['A+', 'A', 'AA+', 'BB+', 'BB', 'BB+', 'AA', 'A+', 'BBB-', 'AA-'],
    'etf': ['EWI', 'EWP', 'EWQ', 'EWZ', 'TUR',
            'EZA', 'EWU', 'FXI', 'ERUS', 'EWJ'],
    'ig_to_hy': [False, False, False, True, False, True, False, False, False, False],
})

print("Sovereign Downgrade Sample")
print("=" * 80)
print(downgrades[['country', 'date', 'old_rating', 'new_rating', 'ig_to_hy']].to_string(index=False))
print(f"\n{len(downgrades)} downgrades, {downgrades['ig_to_hy'].sum()} cross the IG/HY boundary")
```

```python
# -- Download country ETF and benchmark data --
start_date = '2009-01-01'
end_date = '2024-01-01'

all_etfs = downgrades['etf'].tolist()
benchmarks = ['^GSPC', 'SPY']  # S&P 500 as global equity benchmark

all_tickers = list(set(all_etfs + benchmarks))
data = yf.download(all_tickers, start=start_date, end=end_date, progress=False)
if isinstance(data.columns, pd.MultiIndex):
    prices = data['Close']
else:
    prices = data[['Close']]

returns = prices.pct_change().dropna()

# Use SPY as the global benchmark
if 'SPY' in returns.columns:
    bench_ret = returns['SPY']
elif '^GSPC' in returns.columns:
    bench_ret = returns['^GSPC']
else:
    bench_ret = returns.iloc[:, 0]

trading_days = bench_ret.index
print(f"Data: {len(trading_days)} trading days")
print(f"ETFs downloaded: {[t for t in all_etfs if t in returns.columns]}")
missing_etfs = [t for t in all_etfs if t not in returns.columns]
if missing_etfs:
    print(f"Missing: {missing_etfs}")
```

## 3. Computing Abnormal Returns for Sovereign Events

We use the country ETF return as a proxy for the sovereign asset price reaction. The normal return model is a "market model" where the benchmark is the global equity index (SPY). The abnormal return captures the country-specific component of the price change around the downgrade, after removing the global equity market movement.

$$
AR_{i,t} = R_{ETF,i,t} - (\hat{\alpha}_i + \hat{\beta}_i R_{SPY,t})
$$

This specification removes the component of the country ETF return that is driven by global equity market conditions, isolating the country-specific shock. For sovereign bond event studies using actual bond data, one would replace the equity benchmark with a government bond benchmark (e.g., the change in the U.S. 10-year yield or the German Bund yield).

We use a 250-day estimation window with a 10-day buffer and compute CARs for three event windows: $[-1, +1]$, $[-5, +5]$, and $[0, +10]$. The wider post-event window captures any delayed adjustment, which is common in sovereign credit events because bond markets are less liquid than equity markets and because the full implications of a downgrade (index exclusion, collateral haircuts, institutional portfolio rebalancing) take time to materialize.

```python
# -- Estimate market models and compute CARs --
est_length = 250
buffer = 10
event_windows = [(-1, 1), (-5, 5), (0, 10)]

results_list = []

for _, dg in downgrades.iterrows():
    etf = dg['etf']
    edate = dg['date']
    country = dg['country']

    if etf not in returns.columns:
        print(f"  {country} ({etf}): ETF data not available, skipping")
        continue

    etf_ret = returns[etf]
    common = etf_ret.index.intersection(bench_ret.index)
    etf_r = etf_ret.loc[common]
    bench_r = bench_ret.loc[common]
    tdays = etf_r.index

    eidx = tdays.get_indexer([edate], method='ffill')[0]
    ev_s_max = eidx - 5
    ev_e_max = eidx + 10
    est_e = ev_s_max - buffer - 1
    est_s = est_e - est_length + 1

    if est_s < 0 or ev_e_max >= len(tdays):
        print(f"  {country}: insufficient data")
        continue

    y_est = etf_r.iloc[est_s:est_e+1].values
    x_est = bench_r.iloc[est_s:est_e+1].values
    valid = ~(np.isnan(y_est) | np.isnan(x_est))
    if valid.sum() < 60:
        print(f"  {country}: too few valid obs")
        continue

    X = sm.add_constant(x_est[valid])
    ols = sm.OLS(y_est[valid], X).fit()
    alpha, beta = ols.params
    sigma = np.sqrt(ols.mse_resid)

    res = {'country': country, 'etf': etf, 'date': edate,
           'old': dg['old_rating'], 'new': dg['new_rating'],
           'ig_to_hy': dg['ig_to_hy'],
           'alpha': alpha, 'beta': beta, 'sigma': sigma}

    for tau_a, tau_b in event_windows:
        s = eidx + tau_a
        e = eidx + tau_b
        if e >= len(tdays) or s < 0:
            res[f'CAR_[{tau_a:+d},{tau_b:+d}]'] = np.nan
            continue
        y_ev = etf_r.iloc[s:e+1].values
        x_ev = bench_r.iloc[s:e+1].values
        ar = y_ev - (alpha + beta * x_ev)
        res[f'CAR_[{tau_a:+d},{tau_b:+d}]'] = np.nansum(ar)

    results_list.append(res)

sov_df = pd.DataFrame(results_list)

print("\nSovereign Downgrade CARs (%, using country ETF returns)")
print("=" * 85)
print(f"  {'Country':12s} {'Rating':12s} {'[-1,+1]':>9s} {'[-5,+5]':>9s} {'[0,+10]':>9s} {'IG->HY':>7s}")
print(f"  {'-'*60}")
for _, r in sov_df.iterrows():
    vals = [r.get(f'CAR_[{a:+d},{b:+d}]', np.nan) * 100 for a, b in event_windows]
    ig_hy = 'Yes' if r['ig_to_hy'] else ''
    print(f"  {r['country']:12s} {r['old']}->{r['new']:4s} " +
          " ".join(f"{v:+9.2f}" for v in vals) + f"  {ig_hy:>7s}")

caar = sov_df['CAR_[-1,+1]'].mean() * 100
print(f"\n  CAAR [-1,+1]: {caar:+.2f}%")
```

## 4. Testing and Visualization

We apply the standard test battery to the sovereign downgrade sample. Because the sample is small ($N = 10$ or fewer, depending on data availability), the exact binomial sign test is more appropriate than its normal approximation. We also produce the four-panel event study figure for the primary $[-1, +1]$ window.

```python
# -- Tests --
cars = sov_df['CAR_[-1,+1]'].dropna()
N = len(cars)
caar = cars.mean()
se = cars.std(ddof=1) / np.sqrt(N)
t_cs = caar / se if se > 0 else 0
p_cs = 2 * (1 - stats.t.cdf(abs(t_cs), df=max(N - 1, 1)))

n_neg = (cars < 0).sum()  # downgrades should produce negative CARs
t_sign = (n_neg - 0.5 * N) / np.sqrt(0.25 * N)
p_sign = 2 * min(stats.binom.cdf(N - n_neg, N, 0.5),
                 1 - stats.binom.cdf(N - n_neg - 1, N, 0.5))
p_sign = min(p_sign, 1.0)

print("Tests: Sovereign Downgrade CARs [-1, +1]")
print("=" * 55)
print(f"  N = {N}, CAAR = {caar*100:+.2f}%, Negative = {n_neg}/{N}")
print(f"  Cross-sectional t: {t_cs:+.3f} (p = {p_cs:.4f})")
print(f"  Sign test (negative): {t_sign:+.3f} (p = {p_sign:.4f})")

# IG-to-HY subsample
ig_hy = sov_df[sov_df['ig_to_hy'] == True]['CAR_[-1,+1]'].dropna()
non_ig_hy = sov_df[sov_df['ig_to_hy'] == False]['CAR_[-1,+1]'].dropna()
if len(ig_hy) > 0 and len(non_ig_hy) > 0:
    print(f"\n  Subsample analysis:")
    print(f"    IG-to-HY downgrades (N={len(ig_hy)}): CAAR = {ig_hy.mean()*100:+.2f}%")
    print(f"    Within-IG downgrades (N={len(non_ig_hy)}): CAAR = {non_ig_hy.mean()*100:+.2f}%")
    print(f"    Difference: {(ig_hy.mean() - non_ig_hy.mean())*100:+.2f}%")
```

```python
# -- Four-panel figure --

# Build AR panel for the primary window
primary_days = list(range(-1, 2))
ar_panel_sov = pd.DataFrame(index=primary_days, dtype=float)

for _, r in sov_df.iterrows():
    etf = r['etf']
    edate = r['date']
    if etf not in returns.columns:
        continue

    etf_r = returns[etf]
    common = etf_r.index.intersection(bench_ret.index)
    etf_r = etf_r.loc[common]
    bench_r_loc = bench_ret.loc[common]
    tdays = etf_r.index

    eidx = tdays.get_indexer([edate], method='ffill')[0]
    s = eidx - 1
    e = eidx + 1
    if s < 0 or e >= len(tdays):
        continue

    y_ev = etf_r.iloc[s:e+1].values
    x_ev = bench_r_loc.iloc[s:e+1].values
    ar = y_ev - (r['alpha'] + r['beta'] * x_ev)
    ar_panel_sov[r['country']] = ar

if ar_panel_sov.shape[1] > 0:
    N_sov = ar_panel_sov.shape[1]
    aar_sov = ar_panel_sov.mean(axis=1)
    caar_sov = aar_sov.cumsum()
    aar_se = ar_panel_sov.std(axis=1, ddof=1) / np.sqrt(N_sov)
    caar_se = np.sqrt((aar_se ** 2).cumsum())

    fig, axes = plt.subplots(2, 2, figsize=(14, 10))

    ax = axes[0, 0]
    ax.plot(primary_days, caar_sov * 100, 'o-', color='#e31a1c', linewidth=2.5, markersize=6)
    ax.fill_between(primary_days, (caar_sov - 1.96 * caar_se) * 100,
                    (caar_sov + 1.96 * caar_se) * 100, alpha=0.2, color='#e31a1c', label='95% CI')
    ax.axhline(0, color='black', linewidth=0.8)
    ax.axvline(0, color='red', linewidth=1.5, linestyle='--', alpha=0.6, label='Downgrade')
    ax.set_xlabel('Event Day ($\\tau$)'); ax.set_ylabel('CAAR (%)')
    ax.set_title('Panel A: CAAR'); ax.legend(fontsize=9, frameon=True); ax.grid(True, alpha=0.2)

    ax = axes[0, 1]
    c = ['#e31a1c' if v < 0 else '#1f78b4' for v in aar_sov]
    ax.bar(primary_days, aar_sov * 100, color=c, edgecolor='white', width=0.6)
    ax.axhline(0, color='black', linewidth=0.8)
    ax.axvline(0, color='red', linewidth=1.5, linestyle='--', alpha=0.6)
    ax.set_xlabel('Event Day ($\\tau$)'); ax.set_ylabel('AAR (%)')
    ax.set_title('Panel B: AAR'); ax.grid(True, alpha=0.2)

    ax = axes[1, 0]
    for col in ar_panel_sov.columns:
        ax.plot(primary_days, ar_panel_sov[col].cumsum() * 100, 'o-', alpha=0.5, linewidth=1, markersize=4)
    ax.plot(primary_days, caar_sov * 100, 'k-', linewidth=2.5, label='CAAR')
    ax.axhline(0, color='black', linewidth=0.8)
    ax.axvline(0, color='red', linewidth=1.5, linestyle='--', alpha=0.6)
    ax.set_xlabel('Event Day ($\\tau$)'); ax.set_ylabel('CAR (%)')
    ax.set_title('Panel C: Individual CARs'); ax.legend(fontsize=9, frameon=True); ax.grid(True, alpha=0.2)

    ax = axes[1, 1]
    sorted_sov = sov_df.sort_values('CAR_[-1,+1]')
    c2 = ['#e31a1c' if v < 0 else '#1f78b4' for v in sorted_sov['CAR_[-1,+1]']]
    ax.barh(range(len(sorted_sov)), sorted_sov['CAR_[-1,+1]'] * 100, color=c2, edgecolor='white')
    ax.set_yticks(range(len(sorted_sov)))
    ax.set_yticklabels([f"{r['country']} ({r['old']}->{r['new']})" for _, r in sorted_sov.iterrows()], fontsize=8)
    ax.axvline(0, color='black', linewidth=0.8)
    ax.set_xlabel('CAR [-1,+1] (%)'); ax.set_title('Panel D: Cross-Section')
    ax.grid(True, alpha=0.2)

    plt.suptitle('Sovereign Downgrade Event Study', fontsize=14, fontweight='bold', y=1.01)
    plt.tight_layout()
    plt.show()
```

## 5. Corporate Credit Event Studies

### 5.1 From Sovereign to Corporate

Corporate credit event studies follow the same logic as sovereign event studies, with two additional complications. First, the cross-section of corporate bond issuers is much larger, enabling more powerful tests but also requiring careful matching (a BBB-rated industrial firm should not be benchmarked against a AA-rated utility). Second, corporate bond liquidity is even lower than sovereign bond liquidity: many corporate bonds trade only a few times per week.

Bessembinder, Kahle, Maxwell, and Xu (2009) proposed the standard methodology for corporate bond event studies. The key insight is that the abnormal bond return must be computed relative to a matched portfolio of bonds with the same rating and maturity, not relative to the treasury curve. This is because credit spread movements are driven by systematic credit factors (the default risk premium, the liquidity premium) that affect all bonds of the same rating category.

### 5.2 Using CDS as a Proxy

Because daily corporate bond prices are often unavailable or stale, many researchers use CDS spreads instead. CDS contracts are standardized (5-year maturity is the benchmark), trade in dealer markets with daily quotes, and provide a clean measure of the market's assessment of default risk. The CDS spread change $\Delta CDS_{i,t}$ is the fixed income analogue of the equity return $R_{i,t}$.

The normal model for CDS spreads is:

$$
E[\Delta CDS_{i,t}] = \alpha_i + \beta_i \Delta CDS_{index,t}
$$

where $CDS_{index,t}$ is a credit index (CDX IG, CDX HY, or iTraxx). The abnormal spread change is:

$$
\Delta CDS_{i,t}^{abn} = \Delta CDS_{i,t} - (\hat{\alpha}_i + \hat{\beta}_i \Delta CDS_{index,t})
$$

A positive abnormal spread change means the market perceives increased default risk beyond what is explained by market-wide credit conditions. A negative abnormal spread change means the market perceives reduced default risk.

### 5.3 Demonstration: Corporate Rating Downgrades

We demonstrate the approach using equity-based proxies, since free daily CDS data for individual firms is not available via standard APIs. Specifically, we use the stock returns of firms that experienced credit rating downgrades, comparing the equity market reaction to the expected bond market reaction. The equity-credit linkage (Merton, 1974) predicts that a credit downgrade should produce a negative abnormal stock return, with the magnitude depending on the firm's leverage (higher leverage implies a stronger equity-credit link).

```python
# -- Corporate downgrade sample (using equity returns as proxy) --
corp_downgrades = pd.DataFrame({
    'ticker': ['GE', 'F', 'BA', 'T', 'KHC', 'WBA', 'LB', 'M'],
    'company': ['General Electric', 'Ford', 'Boeing', 'AT&T',
                'Kraft Heinz', 'Walgreens', 'L Brands', 'Macys'],
    'downgrade_date': pd.to_datetime([
        '2019-03-14',  # GE: A to BBB+
        '2020-03-25',  # Ford: BBB- to BB+ (fallen angel)
        '2020-01-31',  # Boeing: A to BBB
        '2020-10-23',  # AT&T: BBB+ to BBB
        '2020-02-14',  # Kraft Heinz: BBB- to BB+
        '2020-12-21',  # Walgreens: BBB to BBB-
        '2020-03-27',  # L Brands: BBB- to BB+
        '2020-03-26',  # Macy's: BBB- to BB+
    ]),
    'old_rating': ['A', 'BBB-', 'A', 'BBB+', 'BBB-', 'BBB', 'BBB-', 'BBB-'],
    'new_rating': ['BBB+', 'BB+', 'BBB', 'BBB', 'BB+', 'BBB-', 'BB+', 'BB+'],
    'fallen_angel': [False, True, False, False, True, False, True, True],
})

# Download and estimate
corp_tickers = corp_downgrades['ticker'].tolist()
data_corp = yf.download(corp_tickers + ['SPY'], start='2017-01-01', end='2021-06-01', progress=False)
if isinstance(data_corp.columns, pd.MultiIndex):
    corp_prices = data_corp['Close']
else:
    corp_prices = data_corp[['Close']]
corp_rets = corp_prices.pct_change().dropna()
spy_ret = corp_rets['SPY'] if 'SPY' in corp_rets.columns else corp_rets.iloc[:, -1]
tdays_c = spy_ret.index

corp_results = []
for _, dg in corp_downgrades.iterrows():
    tick = dg['ticker']
    edate = dg['downgrade_date']
    if tick not in corp_rets.columns:
        continue

    eidx = tdays_c.get_indexer([edate], method='ffill')[0]
    ev_s = eidx - 1
    ev_e = eidx + 1
    est_e = ev_s - 10 - 1
    est_s = est_e - 250 + 1
    if est_s < 0 or ev_e >= len(tdays_c):
        continue

    y_est = corp_rets.iloc[est_s:est_e+1][tick].values
    x_est = spy_ret.iloc[est_s:est_e+1].values
    v = ~(np.isnan(y_est) | np.isnan(x_est))
    if v.sum() < 60:
        continue
    X = sm.add_constant(x_est[v])
    ols = sm.OLS(y_est[v], X).fit()
    a, b = ols.params

    y_ev = corp_rets.iloc[ev_s:ev_e+1][tick].values
    x_ev = spy_ret.iloc[ev_s:ev_e+1].values
    ar = y_ev - (a + b * x_ev)
    car = np.nansum(ar)

    corp_results.append({
        'ticker': tick, 'company': dg['company'],
        'old': dg['old_rating'], 'new': dg['new_rating'],
        'fallen_angel': dg['fallen_angel'],
        'CAR': car, 'beta': b,
    })

corp_df = pd.DataFrame(corp_results)

print("Corporate Downgrade CARs [-1,+1] (equity returns)")
print("=" * 70)
for _, r in corp_df.iterrows():
    fa = ' [FALLEN ANGEL]' if r['fallen_angel'] else ''
    print(f"  {r['ticker']:5s} {r['company']:20s} {r['old']}->{r['new']:5s} "
          f"CAR = {r['CAR']*100:+7.2f}%  beta = {r['beta']:.2f}{fa}")

if len(corp_df) > 0:
    print(f"\n  Overall CAAR: {corp_df['CAR'].mean()*100:+.2f}%")
    fa_sub = corp_df[corp_df['fallen_angel']]
    nfa_sub = corp_df[~corp_df['fallen_angel']]
    if len(fa_sub) > 0:
        print(f"  Fallen angels (N={len(fa_sub)}): CAAR = {fa_sub['CAR'].mean()*100:+.2f}%")
    if len(nfa_sub) > 0:
        print(f"  Within-IG (N={len(nfa_sub)}):    CAAR = {nfa_sub['CAR'].mean()*100:+.2f}%")
```

## 6. Methodological Challenges Specific to Fixed Income

### 6.1 Stale Prices and the Bid-Ask Bounce

When a bond has not traded for several days, the reported price is the price of the last trade, which may be stale. Computing daily yield changes from stale prices produces spurious zero returns on non-trading days and a large return on the day of the next trade, which may not coincide with the event. This problem is far more severe than the thin trading problem in equities (Session 2).

Bessembinder et al. (2009) proposed two solutions. First, use only bonds that trade on the event day (this reduces the sample but eliminates staleness). Second, use CDS spreads instead of bond prices (CDS are marked to market daily by dealers even when no trade occurs).

### 6.2 Non-Normality of Spread Changes

Credit spread changes are not normally distributed. They exhibit positive skewness (large spread widenings are more common than large tightenings) and heavy tails. This is because credit risk is inherently asymmetric: a firm can default (large negative event) but cannot "un-default" (there is no symmetric positive event). The non-parametric tests of Session 5 (sign test, rank test) are therefore particularly valuable in fixed income event studies, where the normality assumption underlying the parametric tests is more questionable than in equity markets.

### 6.3 The Maturity Dimension

Bonds and CDS contracts have a term structure of spreads. A credit event may affect the short end (near-term default probability) and the long end (long-run creditworthiness) differently. For example, a liquidity crisis may sharply widen short-term CDS spreads (reflecting immediate default risk) while having a smaller effect on 10-year spreads (if the market expects the firm to survive the near term). A complete fixed income event study should examine multiple maturities.

### 6.4 Contagion

Credit events in fixed income markets are particularly prone to contagion. A sovereign downgrade of Italy widens spreads not only on Italian bonds but also on Spanish, Portuguese, and Greek bonds (Afonso et al., 2012). A corporate default widens spreads on other firms in the same industry. In the standard equity event study, the market model absorbs systematic risk. In fixed income, the credit index benchmark absorbs systematic credit risk, but industry-specific or regional contagion may remain in the abnormal spread change. The researcher should test for contagion explicitly by examining whether the abnormal spread change is positive for non-event firms in the same sector or region.

## 7. Cross-Sectional Determinants of the Downgrade Effect

The literature identifies several determinants of the market reaction to credit rating downgrades.

**Investment-grade to high-yield (fallen angels).** Downgrades that cross the investment-grade boundary (from BBB- to BB+) produce larger abnormal returns than within-category downgrades. This is because many institutional investors (pension funds, insurance companies, bank treasuries) are prohibited from holding high-yield bonds. When a bond is downgraded below investment grade, these investors must sell, creating forced-selling pressure that amplifies the price decline beyond what the credit deterioration alone would justify. This is a well-documented fire-sale effect (Ellul, Jotikasthira, and Lundblad, 2011).

**Number of notches.** A multi-notch downgrade (e.g., from A to BBB) conveys more negative information than a single-notch downgrade and produces a larger market reaction.

**Anticipation.** If the downgrade was preceded by a negative outlook or a credit watch, the market has partially priced in the event, and the announcement-day reaction is smaller. Unanticipated downgrades (no prior watch or outlook change) produce larger CARs.

We test the fallen angel effect using our corporate downgrade sample.

```python
# -- Cross-sectional regression: fallen angel effect --
if len(corp_df) >= 4:
    reg = corp_df.copy()
    reg['CAR_pct'] = reg['CAR'] * 100
    reg['fa_dummy'] = reg['fallen_angel'].astype(int)

    y = reg['CAR_pct']
    X = sm.add_constant(reg[['fa_dummy']])
    fa_reg = sm.OLS(y, X).fit(cov_type='HC1')

    print("Cross-Sectional Regression: Fallen Angel Effect")
    print("=" * 55)
    print(fa_reg.summary2().tables[1].to_string())
    print(f"\nWithin-IG avg CAR:    {fa_reg.params['const']:+.2f}%")
    print(f"Fallen angel premium: {fa_reg.params['fa_dummy']:+.2f}%")
    print(f"  (additional negative CAR for IG-to-HY downgrades)")
else:
    print("Insufficient observations for cross-sectional regression")
```

## 8. Summary

This session extended the event study methodology to fixed income and credit spread markets. The key points are:

The outcome variable in fixed income event studies is the yield change or spread change (in basis points), not the price return. The normal model benchmarks the instrument against a reference sovereign yield (for government bonds) or a matched credit index (for corporate bonds and CDS). CDS spreads are preferable to bond prices when liquidity is a concern, because CDS contracts are standardized and marked to market daily.

Sovereign credit rating downgrades produce significant increases in government bond yields and CDS spreads, with the magnitude depending on the size of the downgrade and whether the country crosses the investment-grade boundary. Corporate downgrades produce significant negative equity returns and positive spread changes, with "fallen angel" downgrades producing the largest effects due to forced selling by institutional investors constrained to investment-grade holdings.

The methodological challenges specific to fixed income, particularly stale prices, non-normality, the maturity dimension, and contagion, require adaptations to the standard equity-based toolkit. Non-parametric tests are especially valuable in this setting, and the researcher should report results for multiple maturities and test for contagion in the cross-section.
