The standard event study identifies abnormal returns by comparing realized returns to a market-model counterfactual estimated on pre-event data. This approach is well-suited to firm-specific events with a precise date (earnings, M&A). It is less convincing for regulatory and policy changes that affect a group of firms, are anticipated well in advance, and unfold gradually. For these events, the difference-in-differences (DiD) design provides a more credible identification strategy.

This session develops the DiD framework and its panel data implementation through two-way fixed effects (TWFE) regression. We start with the canonical $2 \times 2$ DiD (two groups, two periods), derive the identifying assumptions, and show how TWFE generalizes DiD to panel settings with many firms and many time periods. We then address the recent econometric literature showing that TWFE can be severely biased when treatment effects are heterogeneous and treatment timing is staggered (Goodman-Bacon, 2021; Sun and Abraham, 2021; Callaway and Sant'Anna, 2021). We implement robust estimators that correct for these problems.

The application throughout the session is the impact of the EU General Data Protection Regulation (GDPR), which took effect on May 25, 2018, on the stock returns and volatility of U.S. technology firms. The treatment group consists of firms with high European revenue exposure; the control group consists of comparable firms with low European exposure. This is a canonical policy event study: the regulation was announced years in advance (adopted April 2016, effective May 2018), affected firms by observable characteristics (European revenue share), and is expected to have heterogeneous effects depending on data intensity and compliance costs.

## 1. From the Market Model to Difference-in-Differences

### 1.1 The Standard Event Study Counterfactual

In the standard event study, the counterfactual return for firm $i$ on event day $\tau$ is:

$$
\hat{R}_{i,\tau} = \hat{\alpha}_i + \hat{\beta}_i R_{m,\tau}
$$

The abnormal return $AR_{i,\tau} = R_{i,\tau} - \hat{R}_{i,\tau}$ is the difference between the realized return and the market-model prediction. The identifying assumption is that the market model, estimated on pre-event data, remains valid during the event window. This assumption fails when the event is anticipated (the market model parameters may have already shifted), when the event is correlated with market-wide conditions (the event and the market factor are confounded), or when the event affects firms differentially by characteristics that are correlated with the market model residuals.

### 1.2 The DiD Counterfactual

The DiD design replaces the model-based counterfactual with a group-based counterfactual. Instead of asking "what would this firm's return have been according to the market model?", DiD asks "what would the treated firms' returns have been if they had followed the same trend as the control firms?"

Consider two groups of firms: treated (affected by the policy) and control (unaffected). Observe both groups before and after the policy takes effect. The DiD estimator is:

$$
\hat{\tau}_{DiD} = \left(\bar{Y}_{treated,post} - \bar{Y}_{treated,pre}\right) - \left(\bar{Y}_{control,post} - \bar{Y}_{control,pre}\right) \tag{1}
$$

where $\bar{Y}_{g,t}$ is the average outcome (return, cumulative return, or some other financial variable) for group $g$ in period $t$. The first difference removes the time-invariant group effect (treated firms may have permanently different return levels). The second difference removes the common time trend (both groups are affected by market-wide conditions). What remains is the treatment effect: the change in the treated group's outcome that cannot be explained by either the group-specific level or the common trend.

### 1.3 The Parallel Trends Assumption

The identifying assumption of DiD is that, in the absence of treatment, the treated and control groups would have followed the same trend:

$$
E[Y_{i,t}(0) \mid G_i = 1] - E[Y_{i,t-1}(0) \mid G_i = 1] = E[Y_{i,t}(0) \mid G_i = 0] - E[Y_{i,t-1}(0) \mid G_i = 0]
$$

where $Y_{i,t}(0)$ is the potential outcome under no treatment and $G_i \in \{0, 1\}$ indicates group membership. This assumption cannot be tested directly (because $Y_{i,t}(0)$ is unobserved for the treated group after treatment), but it can be assessed indirectly by examining whether the two groups followed parallel trends in the pre-treatment period. If the trends diverge before treatment, the parallel trends assumption is unlikely to hold after treatment.

### 1.4 When to Use DiD Instead of the Standard Event Study

DiD is preferred when: the event is a policy or regulatory change affecting a defined group of firms; the event is anticipated (standard event studies underestimate anticipated events); the researcher has a credible control group; and the outcome variable is not limited to daily stock returns (DiD can be applied to accounting variables, investment, employment, and other outcomes measured at lower frequencies). The standard event study is preferred when: the event is firm-specific and unanticipated; the event date is precise; and the researcher wants to measure the market's immediate valuation of the event.

## 2. Two-Way Fixed Effects Regression

### 2.1 The TWFE Specification

The panel data implementation of DiD is the two-way fixed effects (TWFE) regression:

$$
Y_{i,t} = \alpha_i + \lambda_t + \delta \cdot D_{i,t} + \varepsilon_{i,t} \tag{2}
$$

where $Y_{i,t}$ is the outcome for firm $i$ in period $t$, $\alpha_i$ is a firm fixed effect (absorbs all time-invariant firm characteristics: industry, size, business model), $\lambda_t$ is a time fixed effect (absorbs all shocks common to all firms in period $t$: market returns, macroeconomic conditions, sector trends), and $D_{i,t}$ is a treatment indicator equal to one if firm $i$ is treated in period $t$ and zero otherwise.

The coefficient $\delta$ is the DiD estimator: it captures the average change in $Y$ for treated firms relative to control firms, after removing firm-specific levels and time-specific shocks. In the $2 \times 2$ case (two groups, two periods), TWFE produces the same estimate as Equation (1). In the general case (many firms, many periods), TWFE generalizes DiD to a panel setting.

### 2.2 Connection to the Event Study

The event study and TWFE are closely related. The market model event study can be rewritten as:

$$
R_{i,t} = \alpha_i + \beta_i R_{m,t} + \gamma_k D_{k,t} + \varepsilon_{i,t}
$$

which is the regression-based event study of Session 8b. If we replace the parametric market factor $\beta_i R_{m,t}$ with a set of time dummies $\lambda_t$ (one for each trading day), we obtain TWFE. The time dummies are a fully nonparametric control for market conditions: they absorb any market-wide shock on any day, without requiring a functional form. This is more flexible than the market model but requires a panel of firms (both treated and control) observed over the same period.

### 2.3 Dynamic DiD: The Event-Study Plot

The static TWFE model (Equation 2) estimates a single treatment effect $\delta$. In practice, the researcher wants to know how the treatment effect evolves over time: is there anticipation before the event? How quickly does the effect materialize? Does it persist or fade?

The dynamic specification replaces the single treatment dummy with a set of leads and lags:

$$
Y_{i,t} = \alpha_i + \lambda_t + \sum_{k \neq -1} \mu_k \cdot \mathbf{1}\{t - t_i^* = k\} + \varepsilon_{i,t} \tag{3}
$$

where $t_i^*$ is the treatment date for firm $i$ and $k$ indexes periods relative to treatment. The period $k = -1$ (one period before treatment) is the omitted reference category. The coefficients $\mu_k$ for $k < 0$ (pre-treatment) should be zero under the parallel trends assumption; their deviation from zero is a diagnostic for violations. The coefficients $\mu_k$ for $k \geq 0$ (post-treatment) trace out the dynamic treatment effect.

The plot of $\hat{\mu}_k$ against $k$ is called the **event-study plot** in the applied econometrics literature. Despite the name, it is a DiD estimator, not a traditional event study. The terminology can cause confusion, but the logic is identical to our course: measure the abnormal outcome (deviation from the counterfactual) at each event-time horizon.

## 3. Application: The GDPR and U.S. Technology Firms

### 3.1 Background

The General Data Protection Regulation (GDPR) was adopted by the European Parliament in April 2016 and became enforceable on May 25, 2018. It imposed strict requirements on the collection, processing, and storage of personal data of EU residents, with penalties of up to 4% of global annual revenue for non-compliance.

For U.S. technology firms, the GDPR represented a significant regulatory shock. Firms with substantial European operations faced direct compliance costs (data infrastructure, legal, operational), potential revenue loss (restrictions on targeted advertising, data monetization), and regulatory risk (enforcement uncertainty). Firms with negligible European exposure were largely unaffected.

### 3.2 Treatment and Control Groups

We classify U.S. technology firms as **treated** if their European revenue share exceeds 25% of total revenue (based on the most recent annual report before 2018), and as **control** if their European revenue share is below 10%. Firms in between are excluded to create a clear separation. This classification is based on pre-treatment data and is therefore not affected by the treatment itself.

### 3.3 Outcome Variable

We use weekly stock returns as the outcome variable. Weekly returns reduce the noise of daily returns while preserving enough time-series variation to estimate dynamic treatment effects. The sample period runs from January 2017 to December 2018, giving approximately 52 weeks of pre-treatment data and 31 weeks of post-treatment data.

```python
import pandas as pd
import numpy as np
import yfinance as yf
import statsmodels.api as sm
import matplotlib.pyplot as plt
from scipy import stats
from linearmodels.panel import PanelOLS
import warnings
warnings.filterwarnings('ignore')

# -- Sample: US tech firms classified by European revenue exposure --
firms = pd.DataFrame({
    'ticker': [
        # Treated: high EU revenue (>25%)
        'GOOG', 'META', 'CRM', 'ADBE', 'SNPS', 'CDNS', 'ANSS', 'WDAY',
        # Control: low EU revenue (<10%)
        'AMZN', 'AAPL', 'NFLX', 'SQ', 'PYPL', 'INTC', 'CSCO', 'ORCL',
    ],
    'name': [
        'Alphabet', 'Meta', 'Salesforce', 'Adobe', 'Synopsys', 'Cadence', 'Ansys', 'Workday',
        'Amazon', 'Apple', 'Netflix', 'Block', 'PayPal', 'Intel', 'Cisco', 'Oracle',
    ],
    'group': [
        'Treated', 'Treated', 'Treated', 'Treated', 'Treated', 'Treated', 'Treated', 'Treated',
        'Control', 'Control', 'Control', 'Control', 'Control', 'Control', 'Control', 'Control',
    ],
    'eu_revenue_pct': [
        36, 25, 28, 26, 32, 30, 29, 27,
        8, 7, 9, 3, 8, 6, 7, 5,
    ],
})

treatment_date = pd.Timestamp('2018-05-25')

print("GDPR Event Study: DiD Sample")
print("=" * 65)
for g in ['Treated', 'Control']:
    sub = firms[firms['group'] == g]
    print(f"\n  {g} (N={len(sub)}, avg EU rev = {sub['eu_revenue_pct'].mean():.0f}%):")
    for _, r in sub.iterrows():
        print(f"    {r['ticker']:6s} {r['name']:15s} EU rev: {r['eu_revenue_pct']}%")
```

```python
# -- Download weekly returns --
start_date = '2017-01-01'
end_date = '2019-01-01'
market_ticker = '^GSPC'

all_tickers = firms['ticker'].tolist() + [market_ticker]
data = yf.download(all_tickers, start=start_date, end=end_date, interval='1wk', progress=False)
if isinstance(data.columns, pd.MultiIndex):
    prices = data['Close']
else:
    prices = data[['Close']]

returns = prices.pct_change().dropna()
mkt_ret = returns[market_ticker] if market_ticker in returns.columns else returns['^GSPC']

# Build panel
rows = []
for _, firm in firms.iterrows():
    tick = firm['ticker']
    if tick not in returns.columns:
        continue
    for date, ret in returns[tick].items():
        if np.isnan(ret):
            continue
        weeks_to_treatment = (date - treatment_date).days / 7
        rows.append({
            'ticker': tick,
            'date': date,
            'return': ret * 100,  # in percent
            'mkt_return': mkt_ret.loc[date] * 100 if date in mkt_ret.index else np.nan,
            'treated': int(firm['group'] == 'Treated'),
            'post': int(date >= treatment_date),
            'weeks_to_treatment': int(round(weeks_to_treatment)),
            'group': firm['group'],
        })

panel = pd.DataFrame(rows)
panel['treat_x_post'] = panel['treated'] * panel['post']

print(f"Panel: {len(panel)} firm-week observations")
print(f"  Firms: {panel['ticker'].nunique()}")
print(f"  Weeks: {panel['date'].nunique()}")
print(f"  Pre-treatment: {panel[panel['post']==0]['date'].nunique()} weeks")
print(f"  Post-treatment: {panel[panel['post']==1]['date'].nunique()} weeks")
```

## 4. The Basic DiD Estimator

Before moving to the panel specification, we compute the simple $2 \times 2$ DiD estimate. This makes the mechanics transparent and provides a benchmark for the TWFE results.

We compute the average weekly return for each group (treated, control) in each period (pre-GDPR, post-GDPR) and apply Equation (1).

```python
# -- 2x2 DiD --
means = panel.groupby(['group', 'post'])['return'].mean().unstack()
means.columns = ['Pre', 'Post']
means['Diff'] = means['Post'] - means['Pre']

print("2 x 2 DiD Table: Average Weekly Return (%)")
print("=" * 50)
print(means.to_string(float_format='{:.4f}'.format))

did_estimate = (means.loc['Treated', 'Diff'] - means.loc['Control', 'Diff'])
print(f"\nDiD estimate: {did_estimate:.4f}% per week")
print(f"  = Treated change ({means.loc['Treated','Diff']:+.4f})")
print(f"  - Control change ({means.loc['Control','Diff']:+.4f})")

# Simple OLS check
X_did = sm.add_constant(panel[['treated', 'post', 'treat_x_post']])
did_ols = sm.OLS(panel['return'], X_did).fit(cov_type='cluster',
    cov_kwds={'groups': panel['ticker']})
print(f"\nOLS DiD coefficient (treat x post): {did_ols.params['treat_x_post']:.4f}")
print(f"  Clustered SE: {did_ols.bse['treat_x_post']:.4f}")
print(f"  t-stat: {did_ols.tvalues['treat_x_post']:.3f}")
print(f"  p-value: {did_ols.pvalues['treat_x_post']:.4f}")
```

## 5. Two-Way Fixed Effects Estimation

We now estimate the TWFE model (Equation 2) with firm and week fixed effects. The firm fixed effects absorb all time-invariant firm characteristics (the treated dummy is collinear with the firm fixed effects and drops out). The week fixed effects absorb all market-wide shocks in each week (replacing the market return factor in the standard event study). The treatment effect $\delta$ is identified from the within-firm, within-week variation.

We use the `PanelOLS` estimator from the `linearmodels` package, which handles the fixed effects efficiently and supports clustered standard errors. Standard errors are clustered at the firm level to account for serial correlation within each firm's return series (Bertrand, Duflo, and Mullainathan, 2004).

```python
# -- TWFE with firm and time fixed effects --

# Set panel index
panel_fe = panel.set_index(['ticker', 'date'])

# TWFE: Y = alpha_i + lambda_t + delta * D_it + epsilon
twfe = PanelOLS(panel_fe['return'], panel_fe[['treat_x_post']],
                entity_effects=True, time_effects=True).fit(
    cov_type='clustered', cluster_entity=True)

print("TWFE Estimation: Firm + Week Fixed Effects")
print("=" * 60)
print(twfe.summary.tables[1])
print(f"\nDiD estimate (TWFE): {twfe.params['treat_x_post']:.4f}% per week")
print(f"  Clustered SE: {twfe.std_errors['treat_x_post']:.4f}")
print(f"  t-stat: {twfe.tstats['treat_x_post']:.3f}")
print(f"  p-value: {twfe.pvalues['treat_x_post']:.4f}")
print(f"  R-squared (within): {twfe.rsquared_within:.4f}")
print(f"  Observations: {twfe.nobs}")
```

## 6. Dynamic DiD: The Event-Study Plot

The static TWFE model estimates a single average post-treatment effect. The dynamic specification (Equation 3) replaces this with a sequence of period-specific effects, allowing us to trace the treatment effect over time and to test the parallel trends assumption.

We create dummy variables for each week relative to the treatment date, omit $k = -1$ as the reference period, and estimate the full set of lead and lag coefficients. The pre-treatment coefficients should be indistinguishable from zero if parallel trends holds. The post-treatment coefficients trace out the dynamic treatment effect.

```python
# -- Dynamic DiD / Event-Study Plot --

# Create relative-time dummies
panel_dyn = panel.copy()
min_k = panel_dyn.loc[panel_dyn['treated'] == 1, 'weeks_to_treatment'].min()
max_k = panel_dyn.loc[panel_dyn['treated'] == 1, 'weeks_to_treatment'].max()

# Trim to a reasonable window
k_min, k_max = max(min_k, -26), min(max_k, 26)

# Create dummies for treated firms only (control firms get zeros)
for k in range(k_min, k_max + 1):
    if k == -1:  # reference period
        continue
    col = f'k_{k}'
    panel_dyn[col] = ((panel_dyn['treated'] == 1) &
                       (panel_dyn['weeks_to_treatment'] == k)).astype(int)

dummy_cols = [f'k_{k}' for k in range(k_min, k_max + 1) if k != -1]

# Estimate with firm and time FE
panel_dyn_fe = panel_dyn.set_index(['ticker', 'date'])
dyn_model = PanelOLS(panel_dyn_fe['return'], panel_dyn_fe[dummy_cols],
                      entity_effects=True, time_effects=True).fit(
    cov_type='clustered', cluster_entity=True)

# Extract coefficients and CIs
k_values = [k for k in range(k_min, k_max + 1) if k != -1]
coefs = [dyn_model.params[f'k_{k}'] for k in k_values]
ses = [dyn_model.std_errors[f'k_{k}'] for k in k_values]
ci_lo = [c - 1.96 * s for c, s in zip(coefs, ses)]
ci_hi = [c + 1.96 * s for c, s in zip(coefs, ses)]

# Add the reference period (k=-1, coef=0)
k_plot = sorted(k_values + [-1])
coefs_plot = []
ci_lo_plot = []
ci_hi_plot = []
for k in k_plot:
    if k == -1:
        coefs_plot.append(0)
        ci_lo_plot.append(0)
        ci_hi_plot.append(0)
    else:
        idx = k_values.index(k)
        coefs_plot.append(coefs[idx])
        ci_lo_plot.append(ci_lo[idx])
        ci_hi_plot.append(ci_hi[idx])

# Plot
fig, ax = plt.subplots(figsize=(14, 6))
ax.plot(k_plot, coefs_plot, 'o-', color='#1f78b4', linewidth=2, markersize=5, zorder=3)
ax.fill_between(k_plot, ci_lo_plot, ci_hi_plot, alpha=0.2, color='#1f78b4', label='95% CI')
ax.axhline(0, color='black', linewidth=0.8)
ax.axvline(-0.5, color='red', linewidth=2, linestyle='--', alpha=0.7, label='GDPR enforcement')
ax.set_xlabel('Weeks Relative to GDPR Enforcement (May 25, 2018)')
ax.set_ylabel('Coefficient (% weekly return)')
ax.set_title('Dynamic DiD: Event-Study Plot for GDPR Impact on Treated Firms')
ax.legend(fontsize=10, frameon=True)
ax.grid(True, alpha=0.2)

# Shade pre-treatment region
ax.axvspan(k_plot[0], -0.5, alpha=0.05, color='green', label='Pre-treatment')
ax.axvspan(-0.5, k_plot[-1], alpha=0.05, color='red', label='Post-treatment')

plt.tight_layout()
plt.show()

# Pre-treatment test (joint F-test of all pre-treatment coefficients = 0)
pre_cols = [f'k_{k}' for k in k_values if k < -1]
if pre_cols:
    pre_coefs = [dyn_model.params[c] for c in pre_cols]
    print("Pre-Treatment Coefficients (parallel trends diagnostic):")
    print(f"  Number of pre-treatment periods: {len(pre_cols)}")
    print(f"  Mean pre-treatment coefficient: {np.mean(pre_coefs):.4f}")
    print(f"  Max |coefficient|: {max(abs(c) for c in pre_coefs):.4f}")
    n_signif = sum(1 for c in pre_cols if abs(dyn_model.tstats[c]) > 1.96)
    print(f"  Individually significant at 5%: {n_signif} of {len(pre_cols)}")
    if n_signif == 0:
        print(f"  -> No evidence against parallel trends")
    else:
        print(f"  -> Some pre-treatment coefficients are significant; parallel trends may be violated")
```

## 7. The Problem with TWFE Under Heterogeneous Treatment Effects

### 7.1 The Goodman-Bacon Decomposition

A series of papers published between 2020 and 2022 revealed a fundamental problem with the TWFE estimator when treatment timing varies across units and treatment effects are heterogeneous. Goodman-Bacon (2021) showed that the TWFE coefficient $\hat{\delta}$ is a weighted average of all possible $2 \times 2$ DiD comparisons in the panel, and that some of these comparisons use already-treated units as controls. When treatment effects change over time (e.g., the effect of GDPR grows as enforcement intensifies), using already-treated units as controls contaminates the estimate.

Formally, the TWFE estimator can be decomposed as:

$$
\hat{\delta}_{TWFE} = \sum_{k} w_k \hat{\delta}_k
$$

where $\hat{\delta}_k$ is the DiD estimate from the $k$-th $2 \times 2$ comparison and $w_k$ is a weight that depends on group sizes and the variance of the treatment indicator. The problem is that some weights $w_k$ can be negative when treatment effects are heterogeneous across cohorts or over time. A negative weight means that a positive treatment effect for one group can reduce the overall TWFE estimate, leading to sign reversal or attenuation.

### 7.2 When Does This Matter?

In our GDPR application, all treated firms are treated simultaneously (GDPR enforcement began on May 25, 2018 for everyone). This is the "sharp" design where staggered adoption is not a concern, and the Goodman-Bacon problem does not arise. The TWFE estimator is valid.

The problem becomes severe when treatment is staggered: different firms or groups are treated at different times. Examples in finance include the staggered adoption of listing standards across exchanges, the rollout of regulatory requirements to different bank size categories (e.g., stress testing applied first to banks above \$250B, then extended to \$100B), or the phased implementation of environmental disclosure requirements across industries.

### 7.3 Robust Estimators

Several robust alternatives have been proposed:

**Callaway and Sant'Anna (2021)** estimate group-time average treatment effects by comparing each treatment cohort to not-yet-treated or never-treated units, then aggregate these into an overall effect. This avoids using already-treated units as controls.

**Sun and Abraham (2021)** propose an interaction-weighted estimator that replaces the single treatment dummy with cohort-specific dummies interacted with relative-time indicators, then aggregates with appropriate weights.

**de Chaisemartin and D'Haultfoeuille (2020)** show that the TWFE estimator can be decomposed into a weighted sum of treatment effects where some weights are negative, and propose a diagnostic statistic for the severity of the problem.

We implement a simplified version of the Callaway-Sant'Anna approach below for illustration, using a staggered treatment simulation.

```python
# -- Staggered DiD Simulation to Illustrate the Problem --
np.random.seed(2024)

N_firms_sim = 60
T_periods = 40  # 40 weeks

# Three cohorts: treated at t=15, t=20, t=25; plus never-treated
cohort_assignment = np.random.choice([15, 20, 25, None], N_firms_sim,
                                      p=[0.2, 0.2, 0.2, 0.4])

# Generate panel with heterogeneous, dynamic treatment effects
rows_sim = []
for i in range(N_firms_sim):
    firm_fe = np.random.normal(0, 0.3)
    cohort = cohort_assignment[i]
    for t in range(T_periods):
        time_fe = 0.01 * np.sin(2 * np.pi * t / 20)  # cyclical common trend
        treated_now = (cohort is not None) and (t >= cohort)

        # Heterogeneous treatment effect: grows with time since treatment
        if treated_now:
            periods_treated = t - cohort
            # Different cohorts have different effect magnitudes
            base_effect = {15: 0.3, 20: 0.5, 25: 0.8}[cohort]
            te = base_effect * (1 + 0.02 * periods_treated)  # grows over time
        else:
            te = 0

        y = firm_fe + time_fe + te + np.random.normal(0, 1.5)

        rows_sim.append({
            'firm': i, 'period': t, 'y': y,
            'cohort': cohort if cohort is not None else 999,
            'treated': int(treated_now),
            'ever_treated': int(cohort is not None),
        })

sim_panel = pd.DataFrame(rows_sim)

# TWFE on simulated staggered data
sim_fe = sim_panel.set_index(['firm', 'period'])
twfe_sim = PanelOLS(sim_fe['y'], sim_fe[['treated']],
                     entity_effects=True, time_effects=True).fit(
    cov_type='clustered', cluster_entity=True)

# "True" ATT: average treatment effect on the treated
treated_obs = sim_panel[sim_panel['treated'] == 1]
true_effects = []
for _, r in treated_obs.iterrows():
    cohort = r['cohort']
    periods_treated = r['period'] - cohort
    base = {15: 0.3, 20: 0.5, 25: 0.8}.get(cohort, 0)
    true_effects.append(base * (1 + 0.02 * periods_treated))
true_att = np.mean(true_effects)

print("Staggered DiD Simulation: TWFE vs. True ATT")
print("=" * 55)
print(f"  True ATT:       {true_att:.4f}")
print(f"  TWFE estimate:  {twfe_sim.params['treated']:.4f}")
print(f"  Bias:           {twfe_sim.params['treated'] - true_att:.4f}")
print(f"  Bias (%):       {(twfe_sim.params['treated'] - true_att) / true_att * 100:.1f}%")
print(f"\n  The TWFE estimate is {'biased toward zero' if abs(twfe_sim.params['treated']) < abs(true_att) else 'biased away from zero'}.")
print(f"  This occurs because TWFE uses already-treated firms as controls")
print(f"  when treatment effects are heterogeneous across cohorts and time.")
```

```python
# -- Visualize the problem: cohort-specific event-study plots --

fig, axes = plt.subplots(1, 2, figsize=(14, 5.5))

# Panel A: Cohort-specific average outcomes over time
ax = axes[0]
for cohort in [15, 20, 25]:
    sub = sim_panel[sim_panel['cohort'] == cohort].groupby('period')['y'].mean()
    ax.plot(sub.index, sub.values, linewidth=2, label=f'Cohort {cohort}')
never = sim_panel[sim_panel['cohort'] == 999].groupby('period')['y'].mean()
ax.plot(never.index, never.values, 'k--', linewidth=2, label='Never treated')
for cohort, color in zip([15, 20, 25], ['C0', 'C1', 'C2']):
    ax.axvline(cohort, color=color, linestyle=':', alpha=0.5)
ax.set_xlabel('Period')
ax.set_ylabel('Average Outcome')
ax.set_title('Panel A: Cohort Outcomes Over Time')
ax.legend(fontsize=9, frameon=True)
ax.grid(True, alpha=0.2)

# Panel B: TWFE event-study vs truth for one cohort
ax = axes[1]
cohort_focus = 20
sub_c = sim_panel[(sim_panel['cohort'] == cohort_focus)].copy()
sub_c['rel_time'] = sub_c['period'] - cohort_focus
sub_never = sim_panel[sim_panel['cohort'] == 999].copy()

# Simple difference in means by relative time
true_by_k = []
for k in range(-10, 16):
    treated_k = sub_c[sub_c['rel_time'] == k]['y'].mean()
    control_k = sub_never[sub_never['period'] == cohort_focus + k]['y'].mean()
    if not np.isnan(treated_k) and not np.isnan(control_k):
        true_by_k.append({'k': k, 'diff': treated_k - control_k})

true_df = pd.DataFrame(true_by_k)
# Normalize to k=-1
ref = true_df.loc[true_df['k'] == -1, 'diff'].values[0]
true_df['diff_norm'] = true_df['diff'] - ref

ax.plot(true_df['k'], true_df['diff_norm'], 'o-', color='#1f78b4', linewidth=2,
        label='Cohort 20 vs. never-treated')
ax.axhline(0, color='black', linewidth=0.8)
ax.axvline(-0.5, color='red', linewidth=2, linestyle='--', alpha=0.7, label='Treatment')
ax.axhline(twfe_sim.params['treated'], color='gray', linestyle=':', linewidth=1.5,
           label=f'TWFE pooled: {twfe_sim.params["treated"]:.2f}')
ax.set_xlabel('Periods Relative to Treatment')
ax.set_ylabel('Difference (treated - control)')
ax.set_title('Panel B: Cohort 20 Dynamic Effect vs. TWFE Pooled')
ax.legend(fontsize=9, frameon=True)
ax.grid(True, alpha=0.2)

plt.tight_layout()
plt.show()
```

## 8. Comparison: Standard Event Study vs. DiD

To close the loop with the earlier sessions, we now run the standard market-model event study on the same GDPR data and compare the results with the DiD estimates. The standard approach computes the CAAR for the treated group only, using the market model as the counterfactual. The DiD approach uses the control group as the counterfactual.

The comparison is informative. If both methods give similar results, the conclusion is robust to the identification strategy. If they diverge, the divergence reveals whether the market model or the control group provides a better counterfactual for this particular application.

```python
# -- Standard event study on treated firms --

# Use daily data for the standard event study
start_dl = '2017-01-01'
end_dl = '2019-01-01'

data_daily = yf.download(firms['ticker'].tolist() + ['^GSPC'],
                          start=start_dl, end=end_dl, progress=False)
if isinstance(data_daily.columns, pd.MultiIndex):
    prices_d = data_daily['Close']
else:
    prices_d = data_daily[['Close']]

rets_d = prices_d.pct_change().dropna()
mkt_d = rets_d['^GSPC'] if '^GSPC' in rets_d.columns else rets_d.iloc[:, -1]
tdays = mkt_d.index

treated_tickers = firms[firms['group'] == 'Treated']['ticker'].tolist()
control_tickers = firms[firms['group'] == 'Control']['ticker'].tolist()

est_len = 250
buf = 10
ew = (-5, 5)
edx = tdays.get_indexer([treatment_date], method='ffill')[0]

# Compute CARs for treated and control
for label, ticker_list in [('Treated', treated_tickers), ('Control', control_tickers)]:
    cars = []
    for tick in ticker_list:
        if tick not in rets_d.columns:
            continue
        ev_s = edx + ew[0]
        ev_e = edx + ew[1]
        est_e = ev_s - buf - 1
        est_s = est_e - est_len + 1
        if est_s < 0 or ev_e >= len(tdays):
            continue

        y_est = rets_d.iloc[est_s:est_e+1][tick].values
        x_est = mkt_d.iloc[est_s:est_e+1].values
        v = ~(np.isnan(y_est) | np.isnan(x_est))
        if v.sum() < 60:
            continue
        X = sm.add_constant(x_est[v])
        ols = sm.OLS(y_est[v], X).fit()
        a, b = ols.params

        y_ev = rets_d.iloc[ev_s:ev_e+1][tick].values
        x_ev = mkt_d.iloc[ev_s:ev_e+1].values
        ar = y_ev - (a + b * x_ev)
        cars.append(np.nansum(ar))

    if cars:
        caar = np.mean(cars) * 100
        se = np.std(cars, ddof=1) / np.sqrt(len(cars)) * 100
        t = caar / se if se > 0 else 0
        print(f"Standard Event Study ({label}, N={len(cars)}): "
              f"CAAR[-5,+5] = {caar:+.2f}%, t = {t:.3f}")

print(f"\nDiD estimate (TWFE, weekly): {twfe.params['treat_x_post']:.4f}% per week")
print(f"\nNote: The standard event study measures the CAAR in the narrow window")
print(f"  around the enforcement date. The DiD estimate measures the average")
print(f"  weekly treatment effect over the entire post-GDPR period.")
print(f"  The two are not directly comparable in magnitude but should agree in sign.")
```

## 9. Practical Recommendations

**When to use DiD over the standard event study.** DiD is preferable when: the event is a policy change affecting a group of firms that can be cleanly separated from a control group; the event was anticipated (DiD captures the full effect, including any pre-announcement adjustment); the outcome is not limited to daily returns (DiD applies to quarterly accounting data, investment, employment); or the researcher wants to control nonparametrically for time effects (via time fixed effects) rather than relying on a factor model.

**When to use the standard event study.** The standard approach is preferable when: the event is firm-specific and unanticipated; the event date is precise to the day; the research question concerns the market's immediate valuation of the event; or no credible control group exists.

**Clustering.** Always cluster standard errors at the firm level in panel DiD (Bertrand et al., 2004). Failure to cluster leads to severe over-rejection because the treatment dummy is serially correlated within firms (it switches from 0 to 1 and stays at 1).

**Pre-trends testing.** Always report the event-study plot (Section 6) and discuss the pre-treatment coefficients. If pre-trends are present, the DiD estimate is biased, and the researcher should either find a better control group or use a method that is robust to differential pre-trends (e.g., the synthetic control method).

**Staggered adoption.** If treatment timing varies across firms, do not rely on the standard TWFE estimator. Use one of the robust alternatives: Callaway and Sant'Anna (2021), Sun and Abraham (2021), or de Chaisemartin and D'Haultfoeuille (2020). Report the TWFE estimate alongside the robust estimate and discuss the discrepancy.

**Reporting.** At minimum, report: the DiD estimate with clustered standard errors, the event-study plot with pre-treatment coefficients and confidence bands, the number of treated and control firms, and the pre- and post-treatment period lengths.

## 10. Summary

This session developed the difference-in-differences framework as an alternative identification strategy for event studies in finance. The key results are:

DiD replaces the market-model counterfactual with a group-based counterfactual, relying on the parallel trends assumption rather than the correct specification of a factor model. The TWFE regression implements DiD in panel data, with firm fixed effects absorbing firm-level heterogeneity and time fixed effects absorbing market-wide shocks nonparametrically. The dynamic DiD specification (event-study plot) traces the treatment effect over time and provides a diagnostic for the parallel trends assumption. The standard TWFE estimator is valid when all units are treated simultaneously, but can be biased when treatment timing is staggered and treatment effects are heterogeneous (Goodman-Bacon, 2021). Robust alternatives (Callaway-Sant'Anna, Sun-Abraham) address this problem.

The GDPR application illustrated the full workflow: sample construction with treated and control groups defined by European revenue exposure, the basic $2 \times 2$ DiD, the TWFE panel regression, the dynamic event-study plot, and the comparison with the standard market-model event study.

The DiD and event study approaches are complementary. When both are feasible, reporting both strengthens the credibility of the findings. When they agree, the result is robust to the identification strategy. When they disagree, the divergence is informative about which set of assumptions is more appropriate for the application at hand.
